Western boundary currents develop on the western sides of the ocean basins in response to wind forcing. Poleward flowing western boundary currents like the Gulf Stream and Kuroshio carry vast amounts of heat from the tropics to the mid-latitudes. We will investigate wind-forced ocean circulation and the formation of western boundary currents in an idealised rectangular ocean basin.

A simple depth-independent model of the wind-forced ocean circulation is described by the governing equation for the streamfunction $\psi(x, y, t)$,
\begin{equation}
    \zeta_t + J(\psi, \zeta) + v = -\epsilon\zeta + R\tau
\end{equation}
in $0 \leq x \leq 1$, $0 \leq y \leq 1$ with $\psi = 0$ on the boundaries. Here, $x$ and $y$ are Cartesian coordinates representing eastward and northward directions respectively. The vorticity $\zeta$ is related to $\psi$ through the Poisson equation
\begin{equation}
    \nabla^2\psi = \zeta,
\end{equation}
and the $x$ and $y$ components of the velocity, respectively $u$ and $v$, may be written in terms of $\psi$ as
\begin{equation}
    u = -\frac{\partial\psi}{\partial y}, v = \frac{\partial\psi}{\partial x},
\end{equation}
where $J(\psi, \zeta)$ is the Jacobian with respect to $x$ and $y$ and is an alternative way of writing the advective derivative term $u\cdot\nabla\zeta$. The $-\epsilon\zeta$ term on the right-hand side is a simple representation of
the effect of bottom friction on the flow. The constant $\epsilon$ is a nondimensional frictional damping rate. The term $R\tau(x, y)$, representing the wind forcing, is equal to the curl of the wind stress.

It is convenient to take $\tau$ to be a prescribed function of $x$ and $y$ and, when investigating the behaviour of the model, to vary the strength of the forcing by varying the constant $R$.

The steady state form may be written in the form
\begin{equation}
    u\cdot\nabla(\zeta + y) = R\tau - \epsilon\zeta,
\end{equation}
implying that in the absence of wind-forcing and bottom friction the quantity $\zeta_a = \zeta + y$ would be conserved following the fluid motion. The term $\zeta_a$ is known as the 'absolute vorticity' and is the vertical (i.e. perpendicular to the Earth's surface) component of the vorticity measured with respect to an inertial frame (i.e. including the Earth's rotation as well as the motion of fluid relative to the Earth). The $y$ contribution to the absolute vorticity is a simple representation of the variation of the vertical component of the rotation vector with latitude.

The condition of incompressibility of $u$ is
\begin{equation}
    \nabla \cdot u = 0.
\end{equation}
We use the vector identity for the divergence of a scalar field multiplied by a vector field
\begin{equation}
    \nabla \cdot (u \psi) = (\nabla \cdot u)\psi + u \cdot \nabla \psi.
\end{equation}
Letting $\psi = (\zeta + y)$, incompressibility implies that the first term on the right vanishes, thus
\begin{equation}
    \nabla \cdot u(\zeta + y) = u \cdot \nabla(\zeta + y),
\end{equation}
so the steady state equation can be rewritten as
\begin{equation}
    \nabla \cdot u(\zeta + y) = R\tau - \epsilon\zeta.
\end{equation}
Let $A$ be a region enclosed by a closed streamline $C$. We integrate the rewritten equation over this area $A$
\begin{equation}
    \iint_A \nabla \cdot u(\zeta + y) \, dA = \iint_A (R\tau - \epsilon\zeta) \, dA.
\end{equation}
Now, we apply the divergence theorem to obtain
\begin{equation}
    \oint_C u(\zeta + y) \cdot n \, dl = \iint_A (R\tau - \epsilon\zeta) \, dA.
\end{equation}
By the definition of a streamline, the velocity vector $u$ is everywhere tangent to the curve $C$. Consequently, the velocity is perpendicular to the normal vector $n$ at all points on the boundary, i.e., $u \cdot n = 0$. Therefore, the flux integral on the LHS vanishes, yielding
\begin{equation}
    0 = \iint_A (R\tau - \epsilon\zeta) \, dA.
\end{equation}

If we set $\epsilon = 0$, then the equation simplifies to
\begin{equation}
    0 = R \iint_A \tau \, dA.
\end{equation}
As $\tau$ is either always positive or always negative over the domain and $R$ is a non-zero constant, this integral can never vanish, giving a contradiction, i.e. no steady state solution exists when there is no friction.

We will be concerned with steady-state solutions and their variation as $\epsilon$ and $R$ are changed. However, a convenient way to find the steady-state solution is to integrate the Poisson equation in time, say from initial conditions in which $psi = 0$ for all $x$ and $y$, until the steady state is achieved.
