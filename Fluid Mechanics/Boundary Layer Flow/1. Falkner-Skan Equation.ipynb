Consider flow of an incompressible viscous fluid of constant density $\rho$ and kinematic viscosity $\nu$, whose velocity $u(x, t)$ and hydrodynamic pressure $p(x, t)$ satisfy the incompressibility condition
\begin{equation}
    \nabla \cdot u = 0,
\end{equation}
and the Navier-Stokes momentum equation
\begin{equation}
    \rho (ut + (u\cdot\nabla) u) = -\nabla p + \rho\nu\nabla^2u.
\end{equation}
It follows that the vorticity $\omega \equiv \nabla \times u$ satisfies
\begin{equation}
    \omega_t + (u\cdot\nabla) \omega - (\omega\cdot\nabla) u = \nu\nabla^2\omega.
\end{equation}
If the flow is two-dimensional, then it can be expressed in terms of a streamfunction $\psi(x, y, t)$, with
\begin{equation}
    u â‰¡ (u, v, 0) = \left( \frac{\partial\psi}{\partial y} , -\frac{\partial\psi}{\partial x}, 0 \right), \quad \omega = \left( 0, 0, -\nabla^2\psi\right),
\end{equation}
in terms of which the vorticity equation becomes
\begin{equation}
    \frac{\partial}{\partial t}(\nabla^2\psi) - \frac{\partial (\psi, \nabla^2\psi)}{\partial (x, y)} = \nu\nabla^2(\nabla^2\psi).
\end{equation}
Suppose there is a stationary rigid boundary on $y = 0$, $x > 0$ on which must be satisfied the conditions of no-slip, $u = 0$, and no penetration, $v = 0$. If the viscosity is 'small', then away from the boundary the flow may, to a good approximation, be irrotational: in this case $u \approx \nabla\phi$ for a
velocity potential $\phi$ satisfying $\nabla^2\phi = 0$ (incompressibility) and $\frac{\partial\phi}{\partial y} = 0$ on $y = 0$, $x > 0$ (no penetration).

These conditions, together with corresponding ones on other boundaries and/or
at infinity, are sufficient to determine $\nabla\phi$ uniquely. As a result, it is not possible to specify the tangential slip velocity component on the boundary, i.e.,
\begin{equation}
    \frac{\partial\phi}{\partial x}(x, 0, t) \equiv U_e(x, t), \quad x > 0.
\end{equation}
Instead, $U_e(x, t)$ is determined as part of the irrotational potential-flow solution, and is necessarily non-zero. There must therefore be a boundary layer near $y = 0$ where the potential-flow approximation is not valid and satisfaction of the no-slip condition implies viscous diffusion of vorticity away from the boundary.

If the boundary layer is 'thin', in the sense that within it $\frac{\partial}{\partial x} \ll \frac{\partial}{\partial y}$ (i.e., variations with
respect to $y$ are much more rapid than variations with respect to $x$), then the approximations
\begin{equation}
    \nabla^2\psi \approx \psi_{yy}, \quad \nabla^2(\nabla2\psi) \approx \psi_{yyyy},
\end{equation}
may be made, which can then be integrated once with respect to $y$ to give
\begin{equation}
    \psi_{yt} + \psi_y\psi_{xy} - \psi_x\psi_{yy} = G(x, t) + \nu\psi_{yyy}.
\end{equation}
This is in fact the $x$-component of the Navier-Stokes momentum equation with the $\rho\nu u_{xx}$ term neglected (since it is small compared to $\rho\nu u_{yy}$). The forcing term, or more precisely $-\rho G$, may be identified with the pressure gradient $\frac{\partial p}{\partial x}$ (which in this approximation is independent of $y$, i.e., uniform across the 'thin' boundary layer). This is to be solved in conjunction with the conditions of no slip and no penetration at the boundary, i.e.,
\begin{equation}
    \psi_y = 0, \quad -\psi_x = 0 \implies \psi = \psi_0(t) \quad \text{on} \quad y = 0,
\end{equation}
where $\psi_0(t) = 0$ without loss of generality. In addition the solution must match to the outer potential flow, i.e.,
\begin{equation}
    u \equiv \psi_y \implies U_e(x, t) \quad \text{as} \quad y \to \infty.
\end{equation}
It follows that
\begin{equation}
    G(x, t) = \frac{\partial U_e}{\partial t} + U_e\frac{\partial U_e}{\partial x}.
\end{equation}
In the special case when the flow is steady, $\frac{\partial}{\partial t} = 0$, and when
\begin{equation}
    U_e(x) = Ax^m \quad \text{for} \quad x > 0,
\end{equation}
with $A$ and $m$ constants, the problem admits a similarity solution with
\begin{equation}
    \psi(x, y) = |U_e(x)|\delta(x)f(\eta), \quad \eta = \frac{y}{\delta(x)}, \quad \delta(x) = \left(\frac{\nu x}{|U_e(x)|}\right)^{1/2}.
\end{equation}
Here, $\delta(x)$ is a measure of the local boundary-layer thickness, and the function $f$ satisfies the Falkner-Skan equation
\begin{equation}
    m(f')^2 - \frac{1}{2}(m + 1)ff'' = m + f''',
\end{equation}
with the boundary conditions
\begin{equation}
    f' = f = 0 \quad \text{on} \quad \eta = 0 \quad \text{and} \quad f' \to \mathop{sgn}A \quad \text{as} \quad \eta \to \infty.
\end{equation}
In fact, if there is no source of vorticity other than the boundary $y = 0$, then $f'$ should converge to $\mathop{sgn}A$ exponentially fast as $\eta \to \infty$, i.e.
\begin{equation}
    \eta^N(f' - \mathop{sgn}A) \to 0 \quad \text{as} \quad \eta \to \infty,
\end{equation}
for all $N$. The tangential velocity component is
\begin{equation}
    u = \frac{\partial\psi}{\partial y} = |U_e(x)|f'(\eta).
\end{equation}
The tangential stress (force per unit area in the $x$-direction) exerted by the fluid on the boundary $y = 0$ is
\begin{equation}
    \tau_0 \equiv \rho\nu\left.\frac{\partial u}{\partial y}\right|_{y=0} = \rho\left(\frac{\nu|U_e(x)|^3}{x}\right)^{1/2}f''(0).
\end{equation}


Different values of $m$ arise for different external flows. Particular cases are (taking $A > 0$ unless stated otherwise):
*   $m = 1$: Flow towards a stagnation point on a plane wall.
*   $0 < m < 1$: Flow past a wedge of semi-angle $\theta =\frac{\pi m}{m+1}$.
*   $m = 0$: Flow past a flat plate (the Blasius boundary layer).
*   $-1/2 < m < 0$: Flow around the outside of a corner, turning
through an angle $\theta = -\frac{\pi m}{m+1}$ with slip (but no penetration) upstream of the corner, and no slip (and no penetration) downstream.
*   $m = -1$: Flow due to a line source (for $A > 0$) or line sink ($A < 0$) at the intersection of two plane rigid boundaries (at arbitrary angle).

We ultimately want to solve the two-point-boundary-value problem by shooting,
finding by trial-and-error the values of $f''(0)$ which give a solution with the required behaviour as $\eta \to \infty$. Take constants $B \neq 0$, $k > 0$, $\eta_0$, $k'$ and $k''$.


*   For the case of algebraic convergence, we have the asymptotic behavior:
\begin{align}
    f' &= 1 + B\eta^{-k} + \cdots \quad \text{as} \quad \eta \to \infty, \\
    f &= \eta + C + \frac{B}{1 - k}\eta^{1-k} + \cdots, \\
    f'' &= -kB\eta^{-k-1} + \cdots, \\
    f''' &= k(k + 1)B\eta^{-k-2} + \cdots.
\end{align}
By substituting these into the Falkner-Skan equation, we arrive at the following condition
\begin{equation}
    k = -\frac{m}{1 - m}.
\end{equation}
Therefore the constant that can be determined without reference to the conditions at $\eta = 0$ is $k$ and for this behavior to be possible, we must have $m < 1$.

*   In the case of exponential convergence, the asymptotic form is given by
\begin{align}
    f &= \eta - \eta_0 + e^{-\sigma(\xi)} + \cdots \quad \text{as} \quad \eta \to \infty, \\
    f' &= 1 - \sigma'e^{-\sigma} + \cdots, \\
    f'' &= (\sigma')^2e^{-\sigma} - \sigma''e^{-\sigma} + \cdots, \\
    f''' &= -(\sigma')^3e^{-\sigma} + \cdots,
\end{align}
where $\xi = \eta - \eta_0$ and
\begin{equation}
    \sigma'(\xi) = k\xi + k'+ k''\xi^{-1} + O(\xi^{-2}).
\end{equation}
Substituting these expressions into the Falkner-Skan equation yields
\begin{equation}
    -(\sigma')^3 + \frac{1}{2}(m + 1)(\eta_0 - \xi)(\sigma')^2 - m = 0.
\end{equation}
This leads to the conditions:
\begin{align}
    k &= -1, \\
    k' &= \frac{1}{2}(m + 1)\eta_0, \\
    k'' &= -m - \frac{1}{4}(m + 1)^2\eta_0^2. \\
\end{align}

Thus, the constants $k$, $k'$, and $k''$ can be determined without using the conditions at $\eta = 0$ and this type of behavior is possible for any value of $m$.

*   For algebraic divergence, the solution is given by
\begin{align}
    f &= B\eta^{1+k} + \cdots \quad \text{as} \quad \eta \to \infty, \\
    f' &= (1 + k)B\eta^{k} + \cdots, \\
    f'' &= k(1 + k)B\eta^{k-1} + \cdots, \\
    f''' &= (k - 1)k(1 + k)B\eta^{k-2} + \cdots.
\end{align}
Substituting these into the Falkner-Skan equation gives the leading order equation:
\begin{equation}
    (k - 1)k - \frac{1}{2}(m + 1)k + m = 0.
\end{equation}
The solutions are $k = 1$ or $k = m / 2$. Therefore, the constant that can be determined without conditions at $\eta = 0$ is $k$ and this behavior is possible for any value of $m$.


*   In the case of exponential divergence, we have,
\begin{align}
    f &= Be^{k\eta} + \cdots \quad \text{as} \quad \eta \to \infty, \\
    f' &= kBe^{k\eta} + \cdots, \\
    f'' &= k^2Be^{k\eta} + \cdots, \\
    f''' &= k^3Be^{k\eta} + \cdots.
\end{align}
Substituting into the Falkner-Skan equation, we find that at leading order,
\begin{equation}
    k^3e^{k\eta} - \frac{1}{2}(m + 1)kBe^{\eta}k^2e^{k\eta} = 0.
\end{equation}
As $\eta \to \infty$, the second term dominates, which means there are no solution of this form. Hence, no constants can be determined and this type of behavior is not possible for any $m$.

*   Finally, for a finite-distance singularity, the form is
\begin{align}
    f &= B(\eta_0 - \eta)^{-1} + \cdots \quad \text{as} \quad \eta \to \eta_0, \\
    f' &= B(\eta_0 - \eta)^{-2} + \cdots, \\
    f'' &= 2B(\eta_0 - \eta)^{-3} + \cdots, \\
    f''' &= 6B(\eta_0 - \eta)^{-4} + \cdots.
\end{align}
Substituting into the Falkner-Skan equation leads to
\begin{equation}
    6B - \frac{1}{2}(m + 1)B^2 = 0.
\end{equation}
This gives:
\begin{equation}
    B = \frac{12}{m + 1}.
\end{equation}
Thus, the constant $B$ can be determined without the use of conditions at $\eta$ = 0 and this behavior is possible for any $m  \neq -1$.
