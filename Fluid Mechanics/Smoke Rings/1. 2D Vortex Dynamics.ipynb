A smoke ring is a vortex tube wrapped around into a closed circle (a vortex ring), which propagates normal to the plane of the circle under its self-induced velocity field. We will neglect various effects and crudely model a three-dimensional axisymmetric vortex ring of diameter $a$ and strength $\kappa$ by a pair of point vortices in two-dimensional fluid of strengths $\kappa$ and $-\kappa$, at distance $a$ apart.

The equations of motion for a set of point vortices in a two-dimensional, inviscid, and otherwise stationary fluid are derived from the principle of superposition. Each vortex is advected by the velocity field created by all the other vortices in the system.

A single point vortex is a singularity in the fluid where the vorticity is concentrated at a single point. For a point vortex of strength/circulation $\kappa$ located at the origin, the induced velocity field $u$ at a position $r = (x, y)$ is purely tangential and is given by
\begin{equation}
    u(r) = \frac{\kappa}{2\pi r} \hat\theta
\end{equation}
where $r$ is the distance from the origin and $\hat\theta$ is the unit vector in the tangential direction. In Cartesian coordinates $(x, y)$, this velocity field can be expressed as
\begin{align}
    u(x, y) &= \frac{-\kappa}{2\pi} \frac{y}{r^2} \\
    v(x, y) &=  \frac{\kappa}{2\pi} \frac{x}{r^2}
\end{align}
where $r^2 = x^2 + y^2$.

Now, consider a system of $N$ point vortices. In an inviscid fluid, the flow field generated by multiple vortices is the linear superposition of the fields generated by each individual vortex. Let the $j$-th vortex have strength $\kappa_j$ and be located at position $r_j = (x_j, y_j)$. The velocity induced by this single vortex $j$ at the location of another vortex $i$, which is at position $r_i = (x_i, y_i)$, is found by shifting the coordinate system. The relative position vector is
\begin{equation}
    r_{ij} = r_i - r_j = (x_i - x_j, y_i - y_j).
\end{equation}
The squared distance is
\begin{equation}
    r_{ij}^2 = (x_i - x_j)^2 + (y_i - y_j)^2.
\end{equation}
The velocity components induced at point $i$ by vortex $j$ are given by
\begin{align}
    u_{ij} &= \frac{-\kappa_j}{2\pi} \frac{y_i - y_j}{r_{ij}^2} \\
    v_{ij} &= \frac{\kappa_j}{2\pi} \frac{x_i - x_j}{r_{ij}^2}.
\end{align}
The velocity of the $i$-th vortex is the sum of the velocities induced by all other vortices $j$. Note that a vortex does not move due to its own velocity field. Therefore, we sum the contributions from all other vortices to find the total velocity of vortex $i$
\begin{align}
    \frac{dx_i}{dt} &= \sum_{j \neq i} u_{ij} = \frac{1}{2\pi}\sum_{j \neq i} \frac{-\kappa_j(y_i - y_j)}{r_{ij}^2}, \\
    \frac{dy_i}{dt} &= \sum_{j \neq i} v_{ij} = \frac{1}{2\pi}\sum_{j \neq i} \frac{\kappa_j(x_i - x_j)}{r_{ij}^2},
\end{align}
where $r_{ij} = ((x_i - x_j)^2 + (y_i - y_j)^2)^{1/2}$.


---

The Hamiltonian is defined as
\begin{equation}
H = -\frac{1}{4\pi} \sum_{i \neq j} \kappa_i \kappa_j \log r_{ij}.
\end{equation}
This contains a double summation over all pairs of different vertices $i, j$. When differentiating with respect to the coordinates of a specific vortex $i$, we only need to consider terms that include that vortex. For a given $i$, these terms are $\sum_{j \neq i} \kappa_i \kappa_j \log r_{ij}$ and $\sum_{j \neq i} \kappa_j \kappa_i \log r_{ji}$ which are identical by the symmetry of $r_{ij}$, hence the part of $H$ that depends on $i$ is
\begin{equation}
    H = - \frac{1}{2\pi} \sum_{j \neq i} \kappa_i \kappa_j \log r_{ij}.
\end{equation}
We differentiate $H$ with respect to $x_i$ and $y_i$ using the chain rule,
\begin{align}
    \frac{\partial H}{\partial x_i} &= -\frac{\kappa_i}{2\pi} \sum_{j \neq i} \frac{\kappa_j (x_i - x_j)}{r_{ij}^2}, \\
    \frac{\partial H}{\partial y_i} &= -\frac{\kappa_i}{2\pi} \sum_{j \neq i} \frac{\kappa_j (y_i - y_j)}{r_{ij}^2}.
\end{align}
This shows that
\begin{align}
    \frac{dx_i}{dt} &= \frac{1}{2\pi} \sum_{j \neq i} \frac{-\kappa_j (y_i - y_j)}{r_{ij}^2} = \kappa_i^{-1} \frac{\partial H}{\partial y_i}, \\
    \frac{dy_i}{dt} &= \frac{1}{2\pi} \sum_{j \neq i} \frac{\kappa_j (x_i - x_j)}{r_{ij}^2} = -\kappa_i^{-1} \frac{\partial H}{\partial x_i}.
\end{align}

According to Noether's theorem, each continuous symmetry of the Hamiltonian corresponds to a conserved quantity. The Hamiltonian $H$ depends only on the relative distances $r_{ij}$ between vortices, so it is naturally invariant under translations which shift all vortices by the same amount, and rotations which rotating the entire system of vortices about an origin.

1. Invariance under translation in the $y$-direction, shifting by $(y_i \mapsto y_i + \delta)$. This symmetry implies that $\sum_i \frac{\partial H}{\partial y_i} = 0$. Consider the quantity
\begin{equation}
    I_x = \sum_i \kappa_i x_i.
\end{equation}
Its time derivative is
\begin{equation}
    \frac{dI_x}{dt} = \sum_i \kappa_i \frac{dx_i}{dt} = \sum_i \kappa_i\kappa_i^{-1} \frac{\partial H}{\partial y_i} = \sum_i \frac{\partial H}{\partial y_i} = 0.
\end{equation}
This shows that $I_x$ is conserved which is related to the y-component of the system's linear impulse.

2. Invariance under translation in the $x$-direction, shifting by $(x_i \mapsto x_i + \delta)$. This symmetry implies that $\sum_i \frac{\partial H}{\partial x_i} = 0$. Consider the quantity
\begin{equation}
    I_y = \sum_i \kappa_i y_i.
\end{equation}
Its time derivative is
\begin{equation}
    \frac{dI_y}{dt} = \sum_i \kappa_i \frac{dy_i}{dt} = \sum_i -\kappa_i\kappa_i^{-1} \frac{\partial H}{\partial x_i} = -\sum_i \frac{\partial H}{\partial x_i} = 0.
\end{equation}
This shows that $I_y$ is conserved which is related to the $x$-component of the system's linear impulse.

3. Invariance under rotatation, rotating all position vectors $(x_i, y_i)$ by an angle $\delta$ about the origin. This symmetry implies that $\sum_i (x_i \frac{\partial H}{\partial y_i} - y_i \frac{\partial H}{\partial x_i} = 0$. Consider the quantity
\begin{equation}
    L = \sum_i \kappa_i(x_i^2 + y_i^2).
\end{equation}
Its time derivative is
\begin{align}
    \frac{dL}{dt}
    &= \sum_i \kappa_i\left(2x_i \frac{dx_i}{dt} + 2y_i \frac{dy_i}{dt}\right) \\
    &= 2 \sum_i \kappa_i \left(x_i\kappa_i^{-1} \frac{\partial H}{\partial y_i} - y_i\kappa_i^{-1} \frac{\partial H}{\partial x_i}\right) \\
    &= 2\sum_i \left(x_i \frac{\partial H}{\partial y_i} - y_i \frac{\partial H}{\partial x_i}\right) = 0.
\end{align}
This shows that $L$ is conserved, which is the angular impulse, or the second moment of vorticity.

4. Note that $H$ is also invariant under time translations so the Hamilton itself is conversed. The Hamiltonian is a function of the positions of the vortices $(x_i(t), y_i(t))$. We find its total time derivative is
\begin{align}
    \frac{dH}{dt}
    &= \sum_i \left( \frac{\partial H}{\partial x_i} \frac{dx_i}{dt} + \frac{\partial H}{\partial y_i} \frac{dy_i}{dt} \right) \\
    &= \sum_i \left( \kappa_i^{-1} \frac{\partial H}{\partial x_i} \frac{\partial H}{\partial y_i} - \kappa_i^{-1} \frac{\partial H}{\partial y_i} \frac{\partial H}{\partial x_i} \right) = 0.
\end{align}
As the total time derivative of $H$ is zero, $H$ is a constant of motion. This conserved quantity represents the interaction energy of the vortex system.
