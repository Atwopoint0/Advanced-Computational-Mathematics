For any field $k$, an equation of the form
\begin{equation}
    y^2 + a_1xy + a_3y = x^3 + a_2x^2 + a_4x + a_6
\end{equation}
with coefficients $a_1, a_2, a_3, a_4, a_6 \in k$ is said to define an elliptic curve $E$ over $k$ if the discriminant
\begin{equation}

\end{equation}
is nonzero in the field $k$. Here, the coefficients $b_2, b_4, b_6, b_8$ are given by
\begin{align}
    b_2 &= a^2_1 + 4a_2 \\
    b_4 &= a_1a_3 + 2a_4 \\
    b_6 &= a^2_3 + 4a_6 \\
    b_8 &= a^2_1a_6 - a_1a_3a_4 + 4a_2a_6 + a_2a^2_3 - a^2_4
\end{align}
We use the notation $[a_1, a_2, a_3, a_4, a_6]$ as a shorthand for the above elliptic curve. Geometrically, the condition $\Delta(E) \neq 0 \in k$ ensures that at any solution $(x, y)$ of the equation written in the form $f(x, y) = 0$, there is a well-defined tangent line, meaning that $\frac{\partial f}{\partial x}$ and $\frac{\partial f}{\partial y}$ are not both $0$ at the point $(x, y)$.

For any elliptic curve $E$, let $E(k)$ denote the set of solutions $[x, y, z]$ in the projective plane $\mathbb{P}^2(k)$ to the associated homogeneous equation
\begin{equation}
y^2z + a_1xyz + a_3yz^2 = x^3 + a_2x^2z + a_4xz^2 + a_6z^3.
\end{equation}
Here $\mathbb{P}^2(k)$ is the set of triples $(x, y, z)$ of elements of the field $k$ which are not all $0$, modulo the equivalence relation that $(x, y, z)$ is equivalent to $(cx, cy, cz)$ for $c \in k \setminus\{0\}$.


---

Let $(x_0, y_0)$ be a solution to the affine equation. We can map this solution to a point in the projective plane $\mathbb{P}^2(k)$ by setting the resulting projective point to be $[x_0, y_0, 1]$. To verify that this point lies on the projective curve $E(k)$, note that substituting it into the homogeneous equation yields the affine equation,
\begin{equation}
y_0^2(1) + a_1x_0y_0(1) + a_3y_0(1)^2 = x_0^3 + a_2x_0^2(1) + a_4x_0(1)^2 + a_6(1)^3 \\
y_0^2 + a_1x_0y_0 + a_3y_0 = x_0^3 + a_2x_0^2 + a_4x_0 + a_6.
\end{equation}
This mapping is clearly injective because if we have two distinct affine points $(x_0, y_0) \neq (x_1, y_1)$, then they will map to distinct projective points $[x_0, y_0, 1] \neq [x_1, y_1, 1]$.

Conversely, let $[X, Y, Z]$ be a point in $E(k)$ such that $Z \neq 0$. Due to the equivalence relation in the projective plane, this point is equivalent to $[X/Z, Y/Z, 1]$. Let x = X/Z and y = Y/Z. Since [X, Y, Z] is a solution to the homogeneous equation, we have
\begin{equation}
Y^2Z + a_1XYZ + a_3YZ^2 = X^3 + a_2X^2Z + a_4XZ^2 + a_6Z^3 \\
(Y^2/Z^2) + a_1(X/Z)(Y/Z) + a_3(Y/Z) = (X^3/Z^3) + a_2(X^2/Z^2) + a_4(X/Z) + a_6
\end{equation}
Substituting $x$ and $y$ gives the original affine equation. This proves surjectivity.

Hence we have established a one-to-one correspondence between the set of solutions of the original affine equation and the set of points in $E(k)$ where $z \neq 0$. These points are often referred to as the affine points of the elliptic curve.

The points in $E(k)$ for which $z = 0$ are known as the points at infinity. We substitute z = 0 into the homogeneous equation
\begin{equation}
    y^2(0) + a_1xy(0) + a_3y(0)^2 = x^3 + a_2x^2(0) + a_4x(0)^2 + a_6(0)^3,
\end{equation}
which simplifies significantly to $0 = x^3$, implying that $x = 0$. So any point at infinity is of the form $[0, y, 0]$ for $y \neq 0$ which we can rescale using the equivalence relation to the unique projective point $[0, 1, 0]$.


---

The fundamental fact about elliptic curves is that the set $E(k)$ forms an abelian group in a natural way. The sum of two points $a, b \in E(k)$ is given by drawing the line through $a$ and $b$, which will intersect $E \subset \mathbb{P}^2(k)$ in exactly one other point $c$. The point $c$ will have coefficients in $k$, and the group structure on $E(k)$ is defined by saying that $a + b + c = 0 \in E(k)$.

For a line which is tangent to $E$ at one point $a$ and intersects $E$ at one other point $c$, we interpret the previous equation to mean that $2a + c = 0 \in E(k)$, because we think of the line as intersecting $E$ with multiplicity $2$ at the point $a$. The identity element $0 \in E(k)$ is the point $[0, 1, 0]$.
