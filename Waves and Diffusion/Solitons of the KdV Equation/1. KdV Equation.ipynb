The Korteweg-de Vries (KdV) equation is given by
\begin{equation}
    u_t + uu_x + \delta^2u_{xxx} = 0.
\end{equation}
It arises in many branches of physics as a model for the evolution and interaction of nonlinear waves. It is well-known that the equation possesses single-soliton solutions of the form $u(x, t) = f(x − ct)$ where
\begin{equation}
    f(x) = \frac{A}{\cosh^2 \left(x − x_0/\Delta\right)}, \quad \Delta^2 = 12\delta^2/A, \quad c = A/3,
\end{equation}
and where $A$ and $x_0$ are arbitrary constants representing the amplitude and initial location of the soliton respectively. It is supposed that u(x, t) obeys the cyclic boundary conditions,
\begin{equation}
    u(x + 1, t) = u(x, t),
\end{equation}
so that only the region $0 \leq x \leq 1$ need be considered.

To verify the solution, we must compute the partial derivatives of $u(x,t)$ and substitute them into the KdV equation. Let $z = (x - x0 - ct) / \Delta$. We have
\begin{align}
    u_t &= (2Ac / \Delta) \tanh(z) / \cosh^2(z) \\
    u_x &= (-2A / \Delta) \tanh(z)  / \cosh^2(z) \\
    u_{xxx} &= (-8A / \Delta^3) \tanh(z) / \cosh^2(z) + (24A / \Delta^3) \tanh(z) / \cosh^4(z)
\end{align}
Set $T = \tanh(z)/\cosh^2(z)$ and $S = \tanh(z)/\cosh^2(z)$. Now, substituting these derivatives into the KdV equation yields
\begin{equation}
    (2Ac / \Delta)T + (-2^2A / \Delta)S  + \delta^2 [(-8A / \Delta^3)T + (24A / \Delta^3)S] = 0,
\end{equation}
and grouping by $T$ and $S$ gives
\begin{equation}
[ (2Ac / \Delta) - (8A\delta^2 / \Delta^3) ] T + [ (-2A^2 / \Delta) + (24A\delta^2 / \Delta^3) ] S = 0.
\end{equation}
For this equation to hold true for all values of $z$, the coefficients of both terms must be zero. From the T term,
\begin{equation}
    2Ac / \Delta = 8A\delta^2 / \Delta^3 \implies c = 4δ^2 / Δ^2.
\end{equation}
From the sech^4(z)tanh(z) term,
\begin{equation}
    2A^2 / \Delta = 24A\delta^2 / \Delta^3 \implies A = 12\delta^2 / \Delta^2.
\end{equation}
This is consistent with the conditions given for the coefficients.

The KdV equation is known to have an infinite number of conserved quantities for solutions that decay at infinity or are periodic.

The mass is defined as
\begin{equation}
    M = \int_0^1 u(x, t) \,dx
\end{equation}
To show it is independent of time, we must show that its time derivative is zero. We have
\begin{align}
    \frac{dM}{dt}
    &= \int_0^1 \frac{\partial u}{\partial t} \,dx = \int_0^1 u_t \,dx \\
    &= \int_0^1 (-uu_x - \delta^2 u_{xxx}) \,dx \\
    &= -\int_0^1 uu_x \,dx - \delta^2 \int_0^1 u_{xxx} \,dx
\end{align}
We evaluate each integral. The first integral is
\begin{equation}
    \int_0^1 uu_x \,dx = \int_0^1 \frac{1}{2} \frac{d}{dx}(u^2) \,dx = \left[\frac{1}{2}u^2\right]_0^1 = \frac{1}{2}[u(1,t)^2 - u(0,t)^2].
\end{equation}
Due to the periodic boundary condition $u(1,t) = u(0,t)$, this term is zero. The second integral is
\begin{equation}
    \int_0^1 u_{xxx} \,dx = [u_{xx}]_0^1 = u_{xx}(1,t) - u_{xx}(0,t).
\end{equation}
Since $u$ is periodic, its derivatives are also periodic, so this term is also zero. Therefore, the mass $M$ is conserved over time.

The energy is defined as
\begin{equation}
    E = \int_0^1 \frac{1}{2} u^2(x, t) \,dx.
\end{equation}
Once again, we shall show that its time derivative is zero (dE/dt = 0). We have
\begin{align}
    \frac{dE}{dt}
    &= \int_0^1 \frac{1}{2} 2u \frac{\partial u}{\partial t} \,dx = \int_0^1 uu_t \,dx \\
    &= \int_0^1 u(-uu_x - \delta^2 u_{xxx}) \,dx \\
    &= -\int_0^1 u^2 u_x \,dx - \delta^2 \int_0^1 uu_{xxx} \,dx
\end{align}
The first integral is evaluated as
\begin{equation}
    -\int_0^1 u^2u_x \,dx = -\int_0^1 \frac{1}{3} \frac{d}{dx}(u^3) \,dx = \left[-\frac{1}{3}u^3\right]_0^1 = -\frac{1}{3}[u(1,t)^3 - u(0,t)^3].
\end{equation}
Due to periodicity, this term is zero.
For the second integral, we use parts,
\begin{equation}
    - \delta^2 \int_0^1 uu_{xxx} \,dx = -\delta^2 \left([uu_{xx}]_0^1 - \int_0^1 u_xu_{xx} \,dx \right)
\end{equation}
The boundary term is zero due to periodicity. The integral becomes
\begin{equation}
    \delta^2\int_0^1 u_xu_{xx} \,dx = \delta^2\int_0^1 \frac{1}{2} \frac{d}{dx}(ux)^2 \,dx = δ^2\left[\frac{1}{2}(ux)^2\right]_0^1.
\end{equation}
This term is also zero due to the periodicity of $u_x$. Therefore, the energy $E$ is also conserved over time.
