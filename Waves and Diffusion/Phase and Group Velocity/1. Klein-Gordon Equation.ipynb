The Klein-Gordon equation,
\begin{equation}
    u(x, t): u_{tt} - c_0^2u_{xx} = -q^2u \quad \mbox{with} \quad c_0 > 0 \quad \mbox{and} \quad q > 0
\end{equation}
arises, when looking for solutions of the linear (non-dispersive) wave equation
\begin{equation}
    p_{tt} - c_0^2\nabla^2p = 0
\end{equation}
in the form
\begin{equation}
    p = u(x, t)\cos\left(\frac{m\pi y}{a}\right) \cos\left(\frac{n\pi z}{b}\right)
\end{equation}
to describe the propagation of sound waves in a rectangular tube occupying $0 < y < a$ and $0 < z < b$. It is an example of a wave equation which is both hyperbolic (there is an upper bound on the speed at which waves can propagate) and dispersive (different Fourier components propagate at different speeds).

We wish to solve the Klein-Gordon equation for $x \in \mathbb{R}$ with initial conditions
\begin{equation}
    u(x, 0) = f(x), \quad u_t(x, 0) = 0,
\end{equation}
where $f(x)$ is some specified real function with $f(x) \to 0$ as $|x| \to \infty$, and which for simplicity, we shall assume to be even in $x$, in which case the solution $u(x, t)$ will be even in $x$ for all $t$. Without loss of generality we set $c_0 = 1$.

When $q = 0$, we can reduce to the one-dimensional linear wave equation so that solution is
\begin{equation}
    u(x, t) = \frac{1}{2}(f(x âˆ’ t) + f(x + t)).
\end{equation}

We first apply the Fourier transform to each term of the equation. Using the properties of Fourier transforms for derivatives, we get
\begin{equation}
    \mathcal{F}(u_{xx}) = -k^2\hat{u}(k,t), \quad \mathcal{F}(u_{tt}) = \frac{d^2}{dt^2}\hat{u}(k,t).
\end{equation}
Transforming the entire PDE results in an ODE in $t$,
\begin{equation}
    \frac{d^2}{dt^2} \hat{u}(k,t) + (k^2 + q^2)\hat{u}(k,t) = 0.
\end{equation}
Define the dispersion relation
\begin{equation}
    \Omega(k) = \sqrt(q^2 + k^2).
\end{equation}
Then the equation simplifies to the form of a simple harmonic oscillator
\begin{equation}
    \hat{u}_{tt} + \Omega(k)^2\hat{u} = 0,
\end{equation}
with the general solution
\begin{equation}
    \hat{u}(k,t) = A(k)\cos(\Omega(k)t) + B(k)\sin(\Omega(k)t).
\end{equation}
We use the Fourier transforms of the initial conditions to find the coefficients $A(k)$ and $B(k)$. Define
\begin{equation}
    \tilde{f}(k) = \int_{-\infty}^\infty f(x)e^{-ikx}\,dx.
\end{equation}
From $u(x, 0) = f(x)$, we have $\hat{u}(k, 0) = \tilde{f}(k)$. At $t=0$, the general solution gives $\hat{u}(k, 0) = A(k)$, therefore $A(k) = \tilde{f}(k)$. From $u_t(x, 0) = 0$, we have $\hat{u}_t(k, 0) = 0$. Differentiating the general solution with respect to $t$ gives
\begin{equation}
    \hat{u}_t(k, t) = -A(k)\Omega(k)\sin(\Omega(k)t) + B(k)\Omega(k)\cos(\Omega(k)t).
\end{equation}
At $t=0$, this becomes $\hat{u}_t(k, 0) = B(k)\Omega(k)$, thus $B(k)\Omega(k) = 0$, which implies $B(k) = 0$. The specific solution in the Fourier domain is
\begin{equation}
    \hat{u}(k,t) = \tilde{f}(k)\cos(\Omega(k)t).
\end{equation}
To find $u(x,t)$, we take the inverse Fourier transform,
\begin{align}
    u(x,t)
    &= \frac{1}{2\pi}\int_{-\infty}^\infty \tilde{f}(k)\cos(\Omega(k)t)e^{ikx}\,dk \\
    &= \frac{1}{4\pi}\int_{-\infty}^\infty \tilde{f}(k)(e^{i\Omega(k)t} + e^{-i\Omega(k)t}) e^{ikx}\,dk \\
    &= \frac{1}{4\pi}\int_{-\infty}^\infty \tilde{f}(k)(e^{i(kx + \Omega(k)t)} + e^{i(kx - \Omega(k)t)} )\,dk.
\end{align}
Given that $f(x)$ is a real, even function, its Fourier transform $\tilde{f}(k)$ is also real and even.

The phase velocity $v_p$ and group velocity $v_g$ are defined as
\begin{align}
    v_p(k) = \frac{\Omega(k)}{k} = \frac{\sqrt{q^2 + k^2}}{k}, \\
    v_g(k) = \frac{d\Omega}{dk} = \frac{k}{\sqrt{q^2 + k^2}}.
\end{align}
where $k$ is the wave number. As $k \to 0$, we have $v_p \to \infty$ and $v_g \to 0$. As $k \to \infty$, both $v_p \to 1$ and $v_g \to 1$. Note that the product of the phase and group velocities is a constant $v_pv_g = 1$ since $c_0 = 1$.

Phase velocity is the speed at which a point of constant phase on a wave, such as a crest or a trough, propagates through a medium. For a simple sinusoidal wave, it can be expressed as the ratio of the wavelength to the time period or, more generally, as the ratio of the angular frequency to the angular wave number.

The phase velocity describes the speed of a single frequency component of a wave. If you were to follow a single crest of a simple wave, its speed would be the phase velocity. Crucially, phase velocity does not represent the speed at which information or energy is transmitted. This role is fulfilled by the group velocity, which describes the speed of the overall shape or envelope of a wave packet.

Under certain conditions, the phase velocity can exceed the speed of light in a vacuum. This does not violate the theory of relativity because no information or energy is being transferred faster than $c$.

In a non-dispersive medium, all frequency components of a wave travel at the same speed, meaning the phase velocity equals the group velocity. However, in a dispersive medium, the phase velocity depends on the frequency, causing different components of the wave to travel at different speeds. This causes a wave packet to spread out over time.
