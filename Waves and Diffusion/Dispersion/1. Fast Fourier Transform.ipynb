We shall consider one-dimensional waves, depending on a single spatial coordinate $x$ and time $t$, which are modelled by a system of linear constant coefficient partial differential equations that is second-order in time and time-reversible.

Such a system has single-Fourier-mode 'plane-harmonic-wave' solutions proportional to
\begin{equation}
    e^{ikx \mp i\omega(k)t}
\end{equation}
for any real angular wavenumber $k$, where the angular frequency $\omega$ is real and related to $k$ by a system-dependent dispersion relation. The waves are dispersive if $\omega$ is not directly proportional to $k$. Hence, group velocity $\frac{d\omega}{dk}$ and phase velocity $\frac{\omega}{k}$ vary with $k$, and are unequal.

As an example, one-dimensional capillary-gravity waves on the free surface of
incompressible fluid of uniform depth h have dispersion relation
\begin{equation}
    \omega^2 = (gk + \rho^{-1}\gamma k^3)\tanh{(kh)}
\end{equation}
where $g$ is gravitational acceleration, $\rho$ is the fluid density and $\gamma$ is the coefficient of surface tension.

If the disturbance is described by a function $F(x, t)$, representing say the non-dimensionalised vertical displacement of the fluid surface, the general solution for $F$ will be a superposition of all the Fourier modes,
\begin{equation}
    F(x, t) = \int_{\infty}^\infty (a_+(k)e^{ikx-i\omega(k)t} +
    a_-(k)e^{ikx+i\omega(k)t})\,dk,
\end{equation}
where the amplitudes $a_+(k)$ and $a_-(k)$ are fixed by the initial conditions. For simplicity we shall take these to be
\begin{equation}
    F(x, 0) = e^{-{x^2}/{\sigma^2}}\cos(k_0x) \quad \text{and} \quad \frac{\partial F}{\partial t} (x, 0) = 0,
\end{equation}
where $\sigma$ and $k_0$ are constants.

To apply the second initial condition to the general solution, we need to differentiate $F(x, t)$ with respect to $t$,
\begin{equation}
    \frac{\partial F}{\partial t} = \int_{-\infty}^\infty (-i\omega(k)a_+(k)e^{(ikx - i\omega(k)t)} + i\omega(k)a_-(k)e^{(ikx + i\omega(k)t)}) \,dk.
\end{equation}
Now, we set $t = 0$,
\begin{align}
    \frac{\partial F}{\partial t}(x, 0)
    &= \int_{\infty}^\infty (-i\omega(k)a_+(k)e^{ikx} + i\omega(k)a_-(k)e^{ikx}) \,dk \\
    &= \int_{-\infty}^\infty i\omega(k)e^{ikx} (-a_+(k) + a_-(k)) \,dk = 0.
\end{align}
For this integral to be zero for all values of $x$, the integrand must vanish identically, or equivalently, $a_+(k) = a_-(k)$. Substituting into the original equation yields
\begin{align}
    F(x, t)
    &= \int_{-\infty}^\infty (a_+(k)e^{ikx - i\omega(k)t} + a_+(k)e^{ikx + i\omega(k)t}) \,dk \\
    &= \int_{-\infty}^\infty a_+(k) (e^{ikx - i\omega(k)t} + e^{ikx + i\omega(k)t}) \,dk \\
    &= \int_{-\infty}^\infty a_+(k)(e^{-i\omega(k)t} + e^{i\omega(k)t})e^{ikx} \,dk \\
    &= \int_{-\infty}^\infty 2a_+(k)\cos(\omega(k)t)e^{ikx} \,dk.
\end{align}
Finally, we can define a real function
\begin{equation}
    A(k) = 2a_+(k),
\end{equation}
so that
\begin{equation}
    F(x, t) = \int_{-\infty}^\infty A(k)\cos(\omega(k)t)e^{ikx} \,dk.
\end{equation}
In order to plot the solution some method is needed for evaluating this Fourier integral.



---

The Fourier transform $\hat{G}(k)$ of a function $G(x)$ may be defined by
\begin{equation}
    \hat{G}(k) = \frac{1}{2\pi} \int_{-\infty}^\infty G(x) e^{-ikx} \,dx,
\end{equation}
with inverse
\begin{equation}
    G(x) = \int_{-\infty}^\infty \hat{G}(k) e^{ikx} \,dk .
\end{equation}
The inverse Fourier integral can be approximated by the discretisation
\begin{equation}
    \Delta k\sum_{n = -N/2 + 1}^{N/2} \hat{G}_n e^{in(\Delta k)x}, \quad \hat{G}_n = \hat{G}(n\Delta k),
\end{equation}
provided that $\Delta k$ is small enough to resolve the variation of the integrand with $k$, and that $\hat{G}(k)$ is only significant for $|k| < \frac{1}{2}N\Delta k$. With $\Delta k = 2\pi/L$ and $\Delta x = L/N$, this approximates $G(x = m\Delta x)$ by
\begin{equation}
    g_m \equiv \frac{2\pi}{L} \sum_{n = -N/2 + 1}^{N/2} \hat{G}_n e^{2\pi imn/N} \quad \text{for} \quad -N/2 + 1 \leq m \leq N/2.
\end{equation}
Note that $g_m$ is periodic in $m$ with period $N$, and cannot be expected to give a useful approximation to $G(x = m\Delta x)$ for $|m| > N/2$, i.e., for $|x| > L/2$, since the $e^{ikx}$-factor in the integrand would be chronically under-resolved. This is the exact inverse of
\begin{equation}
    \hat{G}_n = \frac{L}{2\pi N} \sum_{m = -N/2 + 1}{N/2} g_m e^{-2\pi imn/N} \quad \text{for} \quad -N/2 + 1 \leq n \leq N/2;
\end{equation}
which is the right-hand side is a discretisation of the Fourier integral with $k = n\Delta k$, but this will not be required. The discrete Fourier transform and its inverse converge to the Fourier transform and its inverse in the double limit $L \to \infty$ and $N/L \to \infty$.

The fast Fourier transform (FFT) technique is a quick method of evaluating sums of the form
\begin{equation}
    \lambda_m = \sum_{n=0}^{N-1} \mu_n (\zeta_N)^{smn}, \quad m = 0, \dots, N - 1, \quad \zeta_N = e^{2\pi i/N}, \quad s = \pm 1,
\end{equation}
where the $\mu_n$ are a known sequence, and $N$ is a product of small primes, preferably a power of $2$. Note that since
\begin{equation}
    (\zeta_N)^{smn} = (\zeta_N)^{s(m \pm N)n} = (\zeta_N)^{sm(n \pm N)},
\end{equation}
the sums in the discrete Fourier and inverse Fourier transform can be converted to the above FFT form by repositioning part of the series.


```python
import numpy as np
import matplotlib.pyplot as plt

def solve_wave_dispersion(t, g, rho, gamma, h, sigma, k0, N, L):
    '''
    Computes the profile of a dispersive wave packet at a given time t
    using the Fast Fourier Transform method.
    Args:
        t: The time at which to solve for the wave profile.
        g: Gravitational acceleration.
        rho: Fluid density.
        gamma: Surface tension.
        h: Fluid depth.
        sigma: Width of the initial Gaussian envelope.
        k0: Central wavenumber of the initial packet.
        N: Number of grid points (should be a power of 2).
        L: Length of the spatial domain.
    Returns:
        tuple: A tuple containing (x_grid, initial_profile, final_profile).
    '''

    # Spatial grid
    dx = L / N
    x = np.linspace(-L / 2, L / 2, N, endpoint=False)

    # Wavenumber grid (using fftfreq for correct frequency ordering)
    k = 2 * np.pi * np.fft.fftfreq(N, d=dx)

    # Initial condition
    F_x_0 = np.exp(-x**2 / sigma**2) * np.cos(k0 * x)

    # Forward FFT: Transform F(x, 0) to A(k)
    A_k_0 = np.fft.fft(F_x_0)

    # Implement the dispersion relation omega(k)
    omega_sq = (g * np.abs(k) + (gamma / rho) * np.abs(k)**3) * np.tanh(np.abs(k) * h)
    omega = np.sqrt(omega_sq)

    # Evolve the amplitudes in time
    time_evolution_factor = np.cos(omega * t)
    A_k_t = A_k_0 * time_evolution_factor

    # Inverse FFT: Transform back to spatial domain
    F_x_t = np.real(np.fft.ifft(A_k_t))

    return x, F_x_0, F_x_t

time_to_solve = 10.0    # seconds
params = {
    'g': 9.81,          # Gravitational acceleration (m/s^2)
    'rho': 1000,        # Fluid density (kg/m^3)
    'gamma': 0.0728,    # Surface tension of water (N/m)
    'h': 50.0,          # Fluid depth (m)
    'sigma': 10.0,      # Width of the initial Gaussian envelope (m)
    'k0': 2.0,          # Central wavenumber of the initial packet (rad/m)
    'N': 1024,          # Number of grid points
    'L': 200.0          # Length of the spatial domain (m)
}

x, initial_profile, final_profile = solve_wave_dispersion(time_to_solve, **params)
plt.figure(figsize=(8, 5))
plt.plot(x, initial_profile, label='Initial Profile F(x, 0)')
plt.plot(x, final_profile, label=f'Final Profile F(x, t={time_to_solve}s)', linestyle='--')
plt.title('Dispersive Wave Packet Evolution')
plt.xlabel('Position x (m)')
plt.ylabel('Amplitude F(x, t)')
plt.legend()
plt.grid(True)
plt.xlim(-params['L'] / 2, params['L'] / 2)
plt.show()
```
