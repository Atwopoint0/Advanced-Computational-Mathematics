The comoving volume $V_C$ is the volume measure in which the number density of non-evolving objects is constant with redshift. The comoving volume element in solid angle $\sin\theta\,d\theta\,d\phi$ and redshift interval $dz$ is
\begin{equation}
    dV_C = D_H \frac{(1 + z)^2D_A^2}{E(z)}\sin\theta\,dz\,d\theta\,d\phi.
\end{equation}
Integrating this from the present to redshift $z$ gives the total comoving volume over the whole sky to redshift $z$,
\begin{equation}
    V = \frac{4\pi}{3}\frac{D_L^3}{(1 + z)^3} = \frac{4\pi}{3}D_C^3 \quad \text{for} \quad \Omega_k = 0.
\end{equation}
A method to test whether a sample of objects has a uniform comoving density and luminosity which does not change with cosmic time is to use the $\langle V /V_{\max} \rangle$ test. It is assumed that all objects with observed flux $f > f_0$ are detected and the observed flux $f$ and the redshift $z$ are measured for each object.

For a given luminosity $L$ there is a maximum redshift $z_{\max}(L)$ at which the observed flux is $f_0$ so that the object is just included. Corresponding to this redshift is a maximum volume $V_{\max}(L)$. Then, if we have a distribution of luminosities so that $\Phi(L)\,dL$ is the number per unit comoving volume with luminosity between $L$ and $L + dL$, then the total
number of objects in the sample is
\begin{equation}
    \int_0^\infty \Phi(L) \int_0^{V_{\max}(L)} \,dV\,dL.
\end{equation}
where $V$ is the comoving volume.

Assume that we have a uniform comoving distribution so that for a given type of object, the probability of finding it in any given cubic megaparsec of comoving space is the same, regardless of where that volume is located. The expectation value is
\begin{equation}
    \langle V/V_{\max} \rangle = \frac{\int V/V_{\max} \,dN }{\int \,dN },
\end{equation}
where $dN$ is the number of objects in an infinitesimal element of our parameter space. The number of objects $dN$ with luminosity in the range $[L, L+dL]$ found within a comoving volume shell dV is given by
\begin{equation}
    dN = \Phi(L)\,dL\,dV.
\end{equation}
Since $\Phi(L)$ and $V_{\max}(L)$ do not depend on $V$, we can solve the inner integral
\begin{equation}
    âˆ«_0^{V_{\max}(L)}\,dV = [V]_0^{V_{\max}} = V_{\max}(L),
\end{equation}
and therefore the total number of objects in the sample is
\begin{equation}
    \int_0^\infty \Phi(L) V_{\max}(L)\,dL.
\end{equation}
The expression in the numerator is, using the same independence is
\begin{equation}
    \int_0^\infty\Phi(L)/V_{\max}(L)\int_0^{V_{\max}(L)} V\,dV \,dL.
\end{equation}
Solving the inner integral,
\begin{equation}
    \int_0^{V_{\max}(L)} V\,dV = \left[\frac{1}{2}V^2\right]_0^{V_{\max}(L)} = \frac{1}{2}V_{\max}(L)^2,
\end{equation}
which yields the numerator
\begin{equation}
    \int_0^\infty \Phi(L)V_{\max}(L)^2/(2V_{\max}(L))\,dL = \frac{1}{2}\int_0^\infty \Phi(L)V_{\max}(L)\,dL.
\end{equation}
Finally, we obtain
\begin{equation}
    \langle V/V_{\max} \rangle = 1/2.
\end{equation}

```python
import numpy as np
from scipy.integrate import quad
from scipy.optimize import brentq
import io

def parse_data_string(data_string):
    '''
    Parses a multiline string of data into a list of (z, f/f0) tuples.
    '''
    result = [tuple(map(float, line.split())) for line in data_string.strip().splitlines()]
    return result

class CosmologyCalculator:
    '''
    A class to handle cosmological calculations for flat universes.
    '''
    def __init__(self, Omega_m, Omega_Lambda):
        if not np.isclose(Omega_m + Omega_Lambda, 1.0):
            raise ValueError("This class only supports flat universes.")
        self.Omega_m = Omega_m
        self.Omega_Lambda = Omega_Lambda
        self._memo_DL = {}

    def _E_z_integrand(self, z):
        '''
        The integrand 1/E(z) needed for comoving distance.
        '''
        return 1.0 / np.sqrt(self.Omega_m * (1 + z)**3 + self.Omega_Lambda)

    def luminosity_distance(self, z):
        '''
        Calculates luminosity distance DL (normalised by DH) for a given redshift z.
        '''
        if z in self._memo_DL:
            return self._memo_DL[z]

        Dc_div_DH, _ = quad(self._E_z_integrand, 0, z)
        Dl_div_DH = Dc_div_DH * (1 + z) # For a flat universe, DL = DC * (1 + z)
        self._memo_DL[z] = Dl_div_DH
        return Dl_div_DH

    def z_from_luminosity_distance(self, DL_target):
        '''
        Finds the redshift z that corresponds to a target luminosity distance DL.
        '''
        func_to_solve = lambda z: self.luminosity_distance(z) - DL_target
        try:
            return brentq(func_to_solve, a=1e-8, b=20.0) # Search for z in a wide range
        except ValueError:
            return np.nan

    def comoving_volume(self, z):
        '''
        Calculates comoving volume V (normalised).
        '''
        Dl_div_DH = self.luminosity_distance(z)
        Dc_div_DH = Dl_div_DH / (1 + z)
        return Dc_div_DH**3

    def calculate_V_over_Vmax(self, z, flux_ratio):
        '''
        Calculates the V/Vmax ratio for a single object.
        '''
        DL_z = self.luminosity_distance(z)
        DL_target = DL_z * np.sqrt(flux_ratio)
        z_max = self.z_from_luminosity_distance(DL_target)
        if np.isnan(z_max):
            return np.nan
        V = self.comoving_volume(z)
        V_max = self.comoving_volume(z_max)
        return V / V_max if V_max > 0 else np.nan

    def calculate_modified_V_over_Vmax(self, z, flux_ratio, z_min, z_max_survey):
        '''
        Calculates the modified V/Vmax ratio for a survey with redshift limits.
        '''
        # Check if the object is outside the survey limits
        if not (z_min <= z < z_max_survey):
            return np.nan

        # Calculate V_min, a constant for the survey
        V_min = self.comoving_volume(z_min)

        # Calculate V for the object's observed redshift
        V_z = self.comoving_volume(z)

        # Calculate the original z_max based on flux
        DL_z = self.luminosity_distance(z)
        DL_target = DL_z * np.sqrt(flux_ratio)
        z_max_flux = self.z_from_luminosity_distance(DL_target)

        if np.isnan(z_max_flux):
            return np.nan

        # Determine the effective z_max for this object
        z_max_eff = min(z_max_flux, z_max_survey)

        # An object is only in the sample if its potential observable range overlaps
        # with the survey range. z_max_eff must be greater than z_min.
        if z_max_eff <= z_min:
            return np.nan

        # Calculate the max available volume
        V_max_eff = self.comoving_volume(z_max_eff)

        # Calculate the modified ratio
        numerator = V_z - V_min
        denominator = V_max_eff - V_min

        return numerator / denominator if denominator > 0 else np.nan
```

```python
def run_vmax_test(data_string, Omega_m, Omega_Lambda):
    '''
    Parses a data string and calculates the average V/Vmax for a given cosmology.
    '''

    # Parse the input data
    quasar_data = parse_data_string(data_string)

    # Initialise the cosmology calculator
    cosmo_model = CosmologyCalculator(Omega_m=Omega_m, Omega_Lambda=Omega_Lambda)

    # Calculate V/Vmax for each object
    v_vmax_ratios = []
    for z, flux_ratio in quasar_data:
        ratio = cosmo_model.calculate_V_over_Vmax(z, flux_ratio)
        if not np.isnan(ratio):
            v_vmax_ratios.append(ratio)

    # Calculate the average and present the results
    average_V_over_Vmax = np.mean(v_vmax_ratios)
    print(f"The average value of <V/Vmax> is: {average_V_over_Vmax:.4f}")

Omega_m_val = 0.27
Omega_Lambda_val = 0.73
run_vmax_test(quasar_data_string, Omega_m=Omega_m_val, Omega_Lambda=Omega_Lambda_val)
```

A non-evolving population with a constant comoving density would yield $\langle V/V_{\max} \rangle = \frac{1}{2}$, consistent with a uniform distribution. If $\langle V/V_{\max} \rangle > \frac{1}{2}$, then this implies strong cosmic evolution. The quasars are preferentially found farther than expected, meaning they were likely more numerous or more luminous in the past. If $\langle V/V_{\max} \rangle < \frac{1}{2}$, then this implies reverse evolution. The quasars are preferentially found closer than expected, meaning they were less numerous or less luminous in the past.

The obtained value is significantly greater than the expected value of $1/2$. This indicates that the quasars in the sample are not uniformly distributed in comoving volume. They are preferentially found at larger distances (and therefore earlier cosmic times) than a uniform distribution would predict.
This implies strong cosmic evolution. Quasars were either more numerous or more luminous in the past than they are today. The population has evolved significantly over cosmic time.

The original assumes that we can detect objects from a redshift of $z=0$ out to some flux-limited maximum redshift $z_{\max}$. To ensure the expectation value is still $1/2$ for a uniform distribution, we must redefine our volumes to only consider the accessible comoving volume within this window.

```python
def run_modified_vmax_test(data_string, Omega_m, Omega_Lambda, z_min, z_max_survey):
    '''
    Runs the modified V/Vmax test.
    '''

    quasar_data = parse_data_string(data_string)
    cosmo_model = CosmologyCalculator(Omega_m=Omega_m, Omega_Lambda=Omega_Lambda)

    modified_ratios = []
    original_data_count = len(quasar_data)

    # Filter data to only include objects within the survey's redshift range
    filtered_data = [(z, fr) for z, fr in quasar_data if z_min <= z < z_max_survey]

    for z, flux_ratio in filtered_data:
        ratio = cosmo_model.calculate_modified_V_over_Vmax(z, flux_ratio, z_min, z_max_survey)
        if not np.isnan(ratio):
            modified_ratios.append(ratio)

    average_ratio = np.mean(modified_ratios)
    print(f"The modified <Ve/Va> is: {average_ratio:.4f}")

run_modified_vmax_test(
    data_string=quasar_data_string,
    Omega_m=0.27,
    Omega_Lambda=0.73,
    z_min=0.20,
    z_max_survey=3.0
)
```

The result is still significantly greater than the $1/2$ expected for a uniform distribution. This confirms the conclusion from the previous test. The quasar population exhibits strong cosmic evolution, meaning they were more common or more luminous in the early universe.
