We shall investigate the motion of a particle or photon in the Schwarzschild solution to the Einstein field equations.

Use the Lagrangian for a partial in the equatorial place of a Schwarzschild metric
\begin{equation}
    L = F(r)\dot{t}^2 - \frac{\dot{r}^2}{F(r)} - (r\dot\phi)^2,
\end{equation}
where $F(r) = 1 - 1/r$, and we differentiate with respect to a variable $s$ which is an affine parameter along the geodesic. We choose dimensionless units such that $c = 1$ and $2GM = 1$ which makes the Schwarzschild radius $r = 1$. Lagrange's equation for a coordinate $q_i$ is given by
\begin{equation}
    \frac{d}{ds} \frac{\partial L}{\partial \dot{q}_i} - \frac{\partial L}{\partial q_i} = 0.
\end{equation}
We apply Lagrange's equation to the coordinates $t$ and $\phi$. For the t-coordinate, since the Lagrangian does not explicitly depend on $t$, that is $\frac{\partial L}{\partial t} = 0$, the coordinate is cyclic and it implies a conserved quantity,
\begin{equation}
    \frac{\partial L}{\partial \dot{t}} = 2F(r)\dot{t}.
\end{equation}
Lagrange's equation becomes
\begin{equation}
    \frac{d}{ds} (2F(r)\dot{t}) = 0
\end{equation}
This means that $E = F(r)\dot{t}$ is a constant which represents the conserved energy per unit mass. For the \phi -coordinate, similarly, the Lagrangian does not explicitly depend on $\phi$, or $\frac{\partial L}{\partial \phi} = 0$. Hence
\begin{equation}
    \frac{\partial L}{\partial \dot\phi} = -2r^2\dot\phi,
\end{equation}
so Lagrange's equation becomes
\begin{equation}
    \frac{d}{ds} (-2r^2\dot\phi) = 0.
\end{equation}
This implies that $h = r^2\dot\phi$ is also a constant which represents the conserved angular momentum per unit mass.

Instead of using Lagrange's equation for the $r$ coordinate, we can use the Lagrangian itself as a third equation of motion. The value of the Lagrangian $L$ is constant along a geodesic. For massive particles, the affine parameter $s$ can be chosen as the proper time $\tau$, which normalises the Lagrangian to $L = 1$. For photons which travel along null geodesics, we have $L = 0$. The equation is
\begin{equation}
    L = (1 - 1/r)\dot{t}^2 - \frac{\dot{r}^2}{1 - 1/r} - r^2\dot\phi^2.
\end{equation}
Now, we substitute the conserved quantities $E$ and $h$,
\begin{equation}
    L = \frac{E^2}{1 - 1/r} - \frac{\dot{r}^2}{1 - 1/r} - \frac{h^2}{r^2}.
\end{equation}
To get the equation for the orbit shape, we change the variable from $r$ to $u = 1/r$ and the independent variable from $s$ to $\phi$. By the chain rule,
\begin{equation}
    \dot{r} = \frac{dr}{ds} = \frac{dr}{du}\frac{du}{d\phi}\frac{d\phi}{ds} = \frac{-1}{u^2}\frac{du}{d\phi}(hu^2) = -h\frac{du}{d\phi}.
\end{equation}
Substituting $u$ and $\dot{r}$ into the equation for $L$ yields
\begin{equation}
    L = \frac{E^2 - h^2(\frac{du}{d\phi})^2}{1 - u} - h^2u^2,
\end{equation}
which rearranges to
\begin{equation}
    h^2\left(\frac{du}{d\phi}\right)^2 = E^2 - (1 - u)(L + h^2u^2).
\end{equation}

To find the orbit equation involving the second derivative, we differentiate the above expression with respect to $\phi$, using the chain rule $\frac{d}{d\phi} = \frac{du}{d\phi}\frac{d}{du}$,
\begin{align}
    2h^2\frac{du}{d\phi}\frac{d^2u}{d\phi^2}
    &= -\frac{d}{d\phi}((1 - u)(L + h^2u^2)) \\
    &= -\frac{du}{d\phi}\frac{d}{du}(L + h^2u^2 - Lu - h^2u^3) \\
    &= -\frac{du}{d\phi}(2h^2u - L - 3h^2u^2)
\end{align}
Cancelling out the derivative and rearranging gives the general orbit equation,
\begin{equation}
    \frac{d^2u}{d\phi^2} + u = \frac{L}{2h^2} + \frac{3u^2}{2}.
\end{equation}
For massive particles, setting $L = 1$, and recalling that $2GM = 1$, we get
\begin{equation}
    \frac{d^2u}{d\phi^2} + u = \frac{1}{2h^2} + \frac{3u^2}{2}.
\end{equation}
This equation describes the orbits of planets or stars around a non-rotating black hole. The final term is a relativistic correction that accounts for the precession of perihelia.
For massless particles, setting $L = 0$, the equation simplifies to
\begin{equation}
    \frac{d^2u}{d\phi^2} + u = \frac{3u^2}{2}.
\end{equation}
This equation governs the path of light in the Schwarzschild spacetime, which predicts the gravitational lensing of light by massive objects.
