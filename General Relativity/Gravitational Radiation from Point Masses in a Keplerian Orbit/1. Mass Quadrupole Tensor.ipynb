For nearly Newtonian, slow-motion sources in General Relativity, the following are the formulae for the time-averaged losses due to gravitational radiation for the energy and angular momentum of a gravitating source:
\begin{equation}
    \frac{dE}{dt} = -\frac{G}{5c^5}\langle\dddot{I}_{jk}\dddot{I}_{jk}\rangle
\end{equation}
\begin{equation}
    \frac{dJ_j}{dt} = -\frac{2G}{5c^5}\epsilon_{jkl}\langle\ddot{I}_{ki}\dddot{I}_{il}\rangle.
\end{equation}
Here $i$, $j$ and $k$ are the three space indices, the summation convention is used and $\epsilon$ is the alternating tensor. (Note that the coordinate system used is Euclidean, so there is no distinction between upstairs and downstairs indices.) The dot represents a derivative with respect to time $t$. $I_{jk}$ is the trace-free part of the second moment of the mass distribution,
\begin{equation}
    I_{jk} = Q_{jk} - \frac{1}{3}\delta_{jk}Q_{ii},
\end{equation}
where $Q_{jk}$ is the second moment of the mass distribution (the moment of inertia tensor),
\begin{equation}
    Q_{jk} = \sum_a m^{(a)}x^{(a)}_jx^{(a)}_k
\end{equation}
where the sum is over the point masses. For continuous mass distributions the sum is replaced by an integral.
The brackets $\langle\cdot\rangle$ denote an average over a suitably large spacetime $4$-volume.

Consider a Keplerian orbit of two point masses $m_1$ and $m_2$ in the $x$-$y$ plane, with semi-major axis $a$ and eccentricity $e$. Assume the origin is the centre of mass. It is convenient to introduce the total mass $M = m_1 + m_2$ and the reduced mass $\mu = m_1m_2/M$.

Let $d$ be the distance between the two masses and let $\psi$ be the angle of one of the masses relative to the $x$-axis, where the other mass has angle $\psi + \pi$, at a particular point of the orbit; so both $d$ and $\psi$ are functions of time. The first and second point masses are at
\begin{equation}
    d_1(\cos\psi, \sin\psi, 0) \quad \text{and} \quad d_2(-\cos\psi, -\sin\psi, 0),
\end{equation}
where $d_1$ = $d\mu/m_1$ and $d_2 = d\mu/m_2$.


---

Now, we calculate the components of the $Q_jk$ matrix (where indices $1$, $2$, $3$ correspond to $x$, $y$, $z$). We have
\begin{align}
    Q_{11}
    &= m_1(x_1^{(1)})^2 + m_2(x_1^{(2)})^2 \\
    &= m_1(d_1\cos\phi)^2 + m_2(-d_2\cos\psi)^2 \\
    &= (m_1d_1^2 + m_2d_2^2)\cos^2\psi \\
    &= \left(\frac{d^2\mu^2}{m_1} + \frac{d^2\mu^2}{m_2}\right)\cos^2\psi \\
    &= d^2μ^2\left(\frac{1}{m_1} + \frac{1}{m_2}\right)\cos^2\psi \\
    &= d^2μ^2\frac{1}{\mu}\cos^2\psi = d^2\mu\cos^2\psi.
\end{align}
Similarly,
\begin{align}
    Q_{22}
    &= m_1(x_2^{(1)})^2 + m_2(x_2^{(2)})^2 \\
    &= m_1(d_1\sin\psi)^2 + m_2(-d_2\sin\psi)^2 \\
    &= (m_1d_1^2 + m_2d_2^2)\sin^2\psi = d^2\mu\sin^2\psi,
\end{align}
and
\begin{align}
    Q_{12} = Q_{21}
    &= m_1(x_1^{(1)})(x_2^{(1)}) + m_2(x_1^{(2)})(x_2^{(2)}) \\
    &= m_1(d_1\cos\psi)(d_1\sin\psi) + m_2(-d_2\cos\psi)(-d_2\sin\psi) \\
    &= (m_1d_1^2 + m_2d_2^2)\cos\psi\sin\psi = d^2\mu\cos\psi\sin\psi.
\end{align}
Finally, since the motion is in the $x$-$y$ plane, the $z$-coordinates are always zero. Therefore, any component of $Q_{jk}$ involving the index $3$ will be zero.
\begin{equation}
    Q_{33} = Q_{13} = Q_{31} = Q_{23} = Q_{32} = 0.
\end{equation}
In matrix form, this is
\begin{equation}
    Q = d^2\mu
    \begin{pmatrix}
        \cos^2\psi          & \cos\psi\sin\psi  & 0 \\
        \cos\psi\sin\psi    & \sin^2\psi        & 0 \\
        0                   & 0                 & 0
    \end{pmatrix}.
\end{equation}

The trace-free part $I_{jk}$ can be computed from taking the trace of $Q_{jk}$. The diagonal elements are
\begin{equation}
    Q_{ii} = d^2\mu(\cos^2\psi + \sin^2\psi + 0) = d^2\mu.
\end{equation}
This yields
\begin{align}
    I_{11} &= Q_{11} - \delta_{11}Q_{ii}/3 = d^2\mu(\cos^2\psi - 1/3), \\
    I_{22} &= Q_{22} - \delta_{22}Q_{ii}/3 = d^2\mu(\sin^2\psi - 1/3), \\
    I_{33} &= Q_{33} - \delta_{33}Q_{ii}/3 = -d^2\mu/3.
\end{align}
For the off-diagonal elements, the Kronecker delta $\delta_{jk}$ is zero, so $I_{jk} = Q_{jk}$, hence
\begin{equation}
    I_{12} = I_{21} = d^2\mu\cos\psi\sin\psi,
\end{equation}
and
\begin{equation}
    I_{13} = I_{31} = I_{23} = I_{32} = 0.
\end{equation}
Once again, in matrix form
\begin{equation}
    I = d^2\mu
    \begin{pmatrix}
        \cos^2\psi - 1/3    & \cos\psi\sin\psi  & 0 \\
        \cos\psi\sin\psi    & \sin^2\psi - 1/3  & 0 \\
        0                   & 0                 & -1/3
    \end{pmatrix}.
\end{equation}


---

For Keplerian motion the orbit equation is
\begin{equation}
    d =\frac{a(1 - e^2)}{1 + e\cos\psi}
\end{equation}
and the angular velocity is given by
\begin{equation}
    \dot{\psi} = \frac{\sqrt{GM a(1 − e^2)}}{d^2}.
\end{equation}
Assume that $\langle\cdot\rangle$ denotes an average in time over one orbit. For a function $w$ of $\psi$, this means that
\begin{equation}
    \langle w \rangle = \frac{1}{T} \int_0^T w(\psi(t)) \,dt = \frac{1}{T} \int_0^{2\pi} \frac{w(\psi)}{\dot\psi}\,d\psi,
\end{equation}
where
\begin{equation}
    T = \frac{2\pi a^{3/2}}{\sqrt{GM}}
\end{equation}
is the period of the orbit. Assume that $a$ and $e$ do not change appreciably in one orbit, so that their time derivatives can be ignored in the averages. Note that $J_1 = J_2 = 0$.

We shall use a computer algebra system (CAS) to compute the time-average energy and angular momentum.

```python
import sympy
from sympy import symbols, Matrix, cos, sin, sqrt, pi, integrate, Rational, simplify, init_printing
from IPython.display import display, Math, Latex

init_printing(use_latex = True)

def evaluate_time_averages():

    # Define all symbols
    G, M, a, e, mu, c, psi = symbols('G M a e mu c psi', real=True, positive=True)
    # Define the Keplerian orbit expressions
    d = a * (1 - e**2) / (1 + e * cos(psi))
    psi_dot = sqrt(G * M * a * (1 - e**2)) / d**2
    T = 2 * pi * a**(Rational(3, 2)) / sqrt(G * M)

    # Define the trace-free quadrupole moment tensor I_jk
    I_jk = mu * d**2 * Matrix([
        [cos(psi)**2 - Rational(1, 3), cos(psi)*sin(psi), 0],
        [cos(psi)*sin(psi), sin(psi)**2 - Rational(1, 3), 0],
        [0, 0, -Rational(1, 3)]
    ])

    # Define a function for the time derivative d/dt = psi_dot * d/dpsi
    def d_dt(expr):
        # Differentiate with respect to psi and apply the chain rule
        return expr.diff(psi) * psi_dot

    # Calculate the required time derivatives of I_jk
    I_dot = simplify(d_dt(I_jk))
    I_ddot = simplify(d_dt(I_dot))
    I_dddot = simplify(d_dt(I_ddot))

    # Energy time-average
    I_dddot_sq_sum = simplify(sum(comp**2 for comp in I_dddot))
    integrand_E = simplify(I_dddot_sq_sum / psi_dot)
    integral_E = integrate(integrand_E, (psi, 0, 2*pi))
    dE_dt = simplify(-G / (5 * c**5) * (1 / T) * integral_E)
    display(Math(rf'\left\langle\frac{{dE}}{{dt}}\right\rangle = {sympy.latex(dE_dt)}'))

    # Momentum time-average
    term_J = simplify((I_dddot[1,1] - I_dddot[0,0]) * I_ddot[0,1] + (I_ddot[0,0] - I_ddot[1,1]) * I_dddot[0,1])
    integrand_J = simplify(simplify(term_J / psi_dot))
    display(integrand_J)
    integral_J = integrate(integrand_J, (psi, 0, 2*pi))
    dJ3_dt = simplify(- (2 * G) / (5 * c**5) * (1 / T) * integral_J)
    display(Math(rf'\left\langle\frac{{dJ_3}}{{dt}}\right\rangle = {sympy.latex(dJ3_dt)}'))

evaluate_time_averages()
```

We have obtained
\begin{align}
    \left\langle \frac{dE}{dt} \right\rangle
    &= - \frac{G^4\mu^2M^3(37e^4 + 292e^2 + 96)}{15c^5a^5(1-e^2)^{7/2}} \\
    &= - \frac{32G^4\mu^2M^3}{5c^5a^5} \cdot \frac{37e^4 + 292e^2 + 96}{96(1-e^2)^{7/2}} \\
    &= - \frac{32G^4\mu^2M^3}{5c^5a^5} \cdot \frac{1 + \frac{73}{24}e^2 + \frac{37}{96}e^4}{(1-e^2)^{7/2}} \\
    &= - \frac{32G^4\mu^2M^3}{5c^5a^5} f(e),
\end{align}
and
\begin{align}
    \left\langle \frac{dJ_3}{dt} \right\rangle
    &= -\frac{4G^{7/2}\mu^2M^{5/2}(7e^2 + 8)}{5a^{7/2}c^5(e^2-1)^2} \\
    &= -\frac{32G^{7/2}\mu^2M^{5/2}}{5a^{7/2}c^5} \cdot \frac{7e^2 + 8}{8(e^2-1)^2} \\
    &= -\frac{32G^{7/2}\mu^2M^{5/2}}{5c^5a^{7/2}} \cdot \frac{1+\frac{7}{8}e^2}{(1-e^2)^2} \\
    &= -\frac{32G^{7/2}\mu^2M^{5/2}}{5c^5a^{7/2}} g(e).
\end{align}

---

For Keplerian orbits, we have
\begin{equation}
    E = -G \frac{\mu M}{2a}
\end{equation}
and
\begin{equation}
    J_3 = \sqrt{G\mu^2Ma(1-e^2)} =: L.
\end{equation}

Differentiating gives using the chain rule
\begin{equation}
    \frac{dE}{dt} = \frac{dE}{da}\frac{da}{dt} = \frac{G\mu M}{2a^2}\frac{da}{dt},
\end{equation}
then taking the time average gives
\begin{equation}
    \left\langle \frac{dE}{dt} \right\rangle = \left\langle \frac{G\mu M}{2a^2} \frac{da}{dt} \right\rangle.
\end{equation}
The orbital parameters $a$ and $e$ change by a very small amount over a single orbit. Therefore, any function of $a$ and $e$ can be treated as a constant during the integration over one period. This yields
\begin{equation}
    \left\langle\frac{da}{dt}\right\rangle = \left\langle\frac{dE}{dt}\right\rangle \frac{2a^2}{GμM} = - \frac{64G^3\mu M^2}{5c^5a^3} f(e).
\end{equation}
The same idea applies to the angular momentum but now with the multi-variate chain rule,
\begin{equation}
    \frac{dJ_3}{dt} = \frac{\partial J_3}{\partial a}\frac{da}{dt} + \frac{\partial J_3}{\partial e}\frac{de}{dt}.
\end{equation}
Take the time average of the entire equation,
\begin{equation}
    \left\langle\frac{dJ_3}{dt}\right\rangle = \left\langle\frac{\partial J_3}{\partial a}\frac{da}{dt} + \frac{\partial J_3}{\partial e}\frac{de}{dt}\right\rangle = \left\langle\frac{\partial J_3}{\partial a}\frac{da}{dt}\right\rangle + \left\langle\frac{\partial J_3}{\partial e}\frac{de}{dt}\right\rangle.
\end{equation}
Again, under the assumption that $a$ and $e$ change very slowly, the partial derivatives are nearly constant over one orbit and can be treated as constants. Therefore,
\begin{align}
    \left\langle\frac{de}{dt}\right\rangle
    &= \left( \left\langle\frac{dJ_3}{dt}\right\rangle - \frac{\partial J_3}{\partial a} \left\langle\frac{da}{dt}\right\rangle \right) \left(\frac{\partial J_3}{\partial e}\right)^{-1} \\
    &= \left[-\frac{32G^{7/2}\mu^2M^{5/2}}{5c^5a^{7/2}} g(e) + \left(\frac{\sqrt{G\mu^2M(1-e^2)}}{2\sqrt{a}}\frac{64G^3\mu M^2}{5c^5a^3} f(e)\right)\right] \frac{-1}{e} \sqrt{\frac{1-e^2}{G\mu^2Ma}} \\
    &= -\frac{304G^3\mu M^2}{15c^5a^4} \cdot \frac{(1 - \frac{121}{304}e^2)e}{(1-e)^{5/2}} \\
    &= -\frac{304G^3\mu M^2}{15c^5a^4} h(e).
\end{align}

We have now shown that $a$ and $e$ do, in fact, change with time. The assumption is valid if this change is negligible over the timescale of a single orbit, i.e.,
\begin{align}
    |\Delta a| = \left|\left\langle\frac{da}{dt}\right\rangle T\right| \ll a, \\
    |\Delta e| = \left|\left\langle\frac{da}{dt}\right\rangle T\right| \ll e.
\end{align}
The orbital period is $T = 2\pi a^{3/2} / \sqrt{GM}$, so
\begin{equation}
    \frac{|\Delta a|}{a} = \frac{\left|\left\langle\frac{da}{dt}\right\rangle T\right|}{a} = \frac{128\pi G^{5/2}μM^{3/2}}{5c^5a^{5/2}}f(e).
\end{equation}
The term $GM/ac^2$ is a measure of the gravitational potential relative to $c^2$, and is very small in a weak-field (Newtonian) system. The expression is proportional to $(GM/ac^2)^{5/2}$ and since this term is very small, its high power makes the fractional change in a per orbit extremely small. Thus, the time derivatives of $a$ and $e$ can be ignored in the averages if the timescale of the orbit is much shorter than the timescale over which the orbit decays due to gravitational radiation. This condition holds true for nearly Newtonian, slow-motion systems, where orbital velocities are much less than the speed of light.
