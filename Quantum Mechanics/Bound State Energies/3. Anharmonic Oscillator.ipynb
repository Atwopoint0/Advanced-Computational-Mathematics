Now we take the anharmonic oscillator given by
\begin{equation}
    V(x) = \frac{1}{2}x^2 + \frac{b}{6}x^4(x^2 + 1).
\end{equation}

Label the harmonic oscillator eigenstates by $|n\rangle$, i.e. $H_0|n\rangle = E_n|n\rangle$ with
\begin{equation}
    H_0 = \frac{p^2}{2} + \frac{x^2}{2} = a^\dagger a + \frac{1}{2}.
\end{equation}
In dimensionless units, the position operator is given by
\begin{equation}
    x = \frac{1}{\sqrt{2}}(a + a^\dagger).
\end{equation}
To find the operator $x^6$, we must expand the term $(a + a^\dagger)^6$.
\begin{equation}
    x^6 = \frac{1}{2^3} (a + a^\dagger)^6
\end{equation}
This expansion results in a sum of terms, each consisting of exactly six operators. Recall the action of the ladder operators on a state $|n\rangle$:

*   $a^\dagger |n\rangle = \sqrt{n+1} |n+1\rangle$, raises the state index by $1$;
*   $a |n\rangle = \sqrt{n} |n-1\rangle$, lowers the state index by $1$.

In the expansion of $(a + a^\dagger)^6$, the term that raises the state index the most is the term containing only creation operators
\begin{equation}
    (a^\dagger)^6 |n\rangle \propto |n+6\rangle.
\end{equation}
Any other term in the expansion will contain at least one annihilation operator $a$ which lowers the index. Therefore, the highest possible state that can be generated by applying $x^6$ to $|n\rangle$ is $|n+6\rangle$. The matrix element is the inner product, $\langle n+k | x^6 | n \rangle$. As the eigenstates of the harmonic oscillator are orthogonal $\langle m | n \rangle = \delta_{m,n}$, this inner product is non-zero only if the bra state $\langle n+k|$ matches a component of the ket state $x^6|n\rangle$. Since the highest component in $x^6|n\rangle$ is $|n+6\rangle$, the inner product must be zero if the bra index $n+k$ exceeds $n+6$ or $k > 6$.

---

We now wish to derive expressions for the lowest two energies to first order in $b$. To do this, we will require some properties of the Gamma function
\begin{equation}
    \int_{-\infty}^{\infty} z^{2j} e^{-z^2} \,dz = \Gamma\left(j + \frac{1}{2}\right),
\end{equation}
with $\Gamma(1/2) = \sqrt{\pi}$ and $\Gamma(\sigma + 1) = \sigma\Gamma(\sigma)$.

The Hamiltonian is given by
\begin{equation}
    H = H_0 + V' \quad \text{where} \quad V' = \frac{b}{6}(x^4 + x^6).
\end{equation}
The first-order shift in energy is the expectation value of the perturbation in the unperturbed states
\begin{equation}
    E_n^{(1)} = \langle n | V' | n \rangle = \frac{b}{6} \left( \langle n | x^4 | n \rangle + \langle n | x^6 | n \rangle \right).
\end{equation}
The unperturbed ground state wavefunction is
\begin{equation}
    \psi_0(x) = \frac{1}{\pi^{1/4}} e^{-x^2/2} \quad \implies \quad |\psi_0|^2 = \frac{1}{\sqrt{\pi}} e^{-x^2}.
\end{equation}
We now calculate the expectation values $\langle x^m \rangle = \int_{-\infty}^{\infty} x^m |\psi_0|^2 \,dx$:

*   For $x^4$, we have $2j=4$ so $j=2$ and using $\Gamma(5/2) = \frac{3}{4}\sqrt{\pi}$,
\begin{equation}
    \langle 0 | x^4 | 0 \rangle = \frac{1}{\sqrt{\pi}} \int_{-\infty}^{\infty} x^4 e^{-x^2} dx = \frac{1}{\sqrt{\pi}} \Gamma(5/2) = \frac{3}{4}.
\end{equation}

*   For $x^6$, we have $2j=6$ so $j=3$ and using $\Gamma(7/2) = \frac{15}{8}\sqrt{\pi}$,
\begin{equation}
    \langle 0 | x^6 | 0 \rangle = \frac{1}{\sqrt{\pi}} \Gamma(7/2) = \frac{15}{8}.
\end{equation}

Thus, the total energy of the ground state is
\begin{align}
    E_0
    &\approx E_0^{(0)} + \frac{b}{6} \left( \frac{3}{4} + \frac{15}{8} \right) = \frac{1}{2} + \frac{b}{6} \left( \frac{6}{8} + \frac{15}{8} \right) \\
    &\approx \frac{1}{2} + \frac{b}{6} \left( \frac{21}{8} \right) = \frac{1}{2} + \frac{7b}{16}.
\end{align}
The unperturbed first excited state is
\begin{equation}
    \psi_1(x) = \frac{\sqrt{2}}{\pi^{1/4}} x e^{-x^2/2} \quad \implies \quad |\psi_1|^2 = \frac{2}{\sqrt{\pi}} x^2 e^{-x^2}.
\end{equation}
The expectation value becomes
\begin{equation}
    \langle 1 | x^m | 1 \rangle = \int_{-\infty}^{\infty} x^m \left( \frac{2}{\sqrt{\pi}} x^2 e^{-x^2} \right) dx = \frac{2}{\sqrt{\pi}} \int_{-\infty}^{\infty} x^{m+2} e^{-x^2} \,dx.
\end{equation}

*   For $x^4$, we need $\int x^6 e^{-x^2}$, which we found to be $\frac{15}{8}\sqrt{\pi}$,
\begin{equation}
    \langle 1 | x^4 | 1 \rangle = \frac{2}{\sqrt{\pi}} \left( \frac{15}{8}\sqrt{\pi} \right) = \frac{15}{4}.
\end{equation}

*   For $x^6$, here $m+2 = 8$ so $2j=8$ or $j=4$. We need $\Gamma(9/2) = \frac{105}{16}\sqrt{\pi}$, so
\begin{equation}
    \langle 1 | x^6 | 1 \rangle = \frac{2}{\sqrt{\pi}} \left( \frac{105}{16}\sqrt{\pi} \right) = \frac{105}{8}.
\end{equation}
Hence, the total energy of the first excited state is
\begin{align}
    E_1
    &\approx E_1^{(0)} + \frac{b}{6} \left( \frac{15}{4} + \frac{105}{8} \right) = \frac{3}{2} + \frac{b}{6} \left( \frac{30}{8} + \frac{105}{8} \right) \\
    &\approx \frac{3}{2} + \frac{b}{6} \left( \frac{135}{8} \right) = \frac{3}{2} + \frac{45b}{16}.
\end{align}
To summarise, the energies to first order in perturbation theory are
\begin{align}
    E_0 \approx 0.5 + \frac{7}{16}b, \\
    E_1 \approx 1.5 + \frac{45}{16}b.
\end{align}
Without proof, the energies correct to second order in $b$ are
\begin{align}
    E_0 \approx 0.5 + \frac{7}{16}b - \frac{567}{256}b^2, \\    
    E_1 \approx 1.5 + \frac{45}{16}b - \frac{7065}{256}b^2.
\end{align}

```python
import numpy as np
import matplotlib.pyplot as plt

class SchrodingerSolver:
    '''
    A solver for the 1D Schr√∂dinger equation using the matrix method and Given's proceedure.
    '''
    def __init__(self, N, epsilon, potential_func):
        '''
        Initialise the solver with grid parameters and a potential function.
        Args:
            N: Half-number of grid points (total points = 2N).
            epsilon: Grid spacing.
            potential_func: A function V(x) that returns the potential.
        '''
        self.N = N
        self.epsilon = epsilon
        self.num_points = 2 * N

        # Discretise space
        self.x = np.linspace(-N * epsilon, N * epsilon, self.num_points, endpoint=False)

        # Diagonal elements
        self.V = potential_func(self.x)
        self.c = 1.0 + (epsilon**2) * self.V

        # Off-diagonal elements
        self.b_squared = 0.25

    def _count_eigenvalues_below(self, lam):
        '''
        Uses Given's Procedure to count eigenvalues < lambda, i.e., s(lambda).
        '''
        count = 0
        q = self.c[0] - lam

        if q < 0:
            count += 1

        for n in range(1, self.num_points):
            # Handling small q to avoid division by zero
            if abs(q) < 1e-12:
                q = 1e-12
            q = (self.c[n] - lam) - self.b_squared / q

            if q < 0:
                count += 1
        return count

    def find_eigenvalues(self, start_lambda=0.0):
        '''
        Finds the first four eigenvalues starting search from start_lambda.
        Returns a list of Energies (E), not matrix eigenvalues (lambda).
        '''
        energies = []
        current_lambda = start_lambda
        step_size = 0.5 * (self.epsilon**2)

        while len(energies) < 4:
            s_curr = self._count_eigenvalues_below(current_lambda)
            next_lambda = current_lambda + step_size
            s_next = self._count_eigenvalues_below(next_lambda)

            # If count increases, then an eigenvalue is crossed
            if s_next > s_curr:
                # Jumped over a state
                if s_next - s_curr > 1:
                    step_size /= 2
                    continue # Retry this step with smaller size

                # Refine with bisection
                lam_refined = self._bisect(current_lambda, next_lambda, s_curr)

                # Convert matrix lambda to energy E
                E = lam_refined / (self.epsilon**2)
                energies.append(E)

                # Move search forward
                current_lambda = lam_refined + (1e-4 * self.epsilon**2)
            else:
                current_lambda = next_lambda

            # Safety break for infinite loops
            if current_lambda > 100:
                print("Search limit exceeded.")
                break

        return np.array(energies)

    def _bisect(self, lower, upper, target_s_count):
        '''
        Bisection method to refine eigenvalue location.
        '''
        while upper - lower > 1e-12:
            mid = (lower + upper) / 2.0
            if self._count_eigenvalues_below(mid) > target_s_count:
                upper = mid
            else:
                lower = mid
        return (lower + upper) / 2.0

    def compute_eigenvector(self, E, state_index):
        '''
        Computes the eigenvector using symmetry to avoid tail divergence.
        '''
        psi = np.zeros(self.num_points)
        lam = (self.epsilon**2) * E

        # Integrate only from left boundary (n=0) to center (n=N)
        psi[0] = 1.0
        psi[1] = 2 * (self.c[0] - lam) * psi[0]

        # Stop at the center point x=0
        for n in range(1, self.N):
            psi[n+1] = 2 * (self.c[n] - lam) * psi[n] - psi[n-1]

            # Rescale to prevent overflow during calculation
            if abs(psi[n+1]) > 1e10:
                psi[:n+2] /= np.max(np.abs(psi[:n+2]))

        # Mirror the result to the right side around the center index N
        parity = 1 if (state_index % 2 == 0) else -1
        for k in range(1, self.N):
            if self.N + k < self.num_points:
                psi[self.N + k] = parity * psi[self.N - k]

        # Normalise
        norm = np.sqrt(np.sum(psi**2) * self.epsilon)
        psi = psi / norm

        # Convention: The main lobe on the right side (x > 0) is positive.
        right_half = psi[self.N:]
        # If the main feature on the right is negative, flip the whole wavefunction
        max_val_right = right_half[np.argmax(np.abs(right_half))]
        if max_val_right < 0:
            psi *= -1.0

        return psi

def plot_quantum_states(x, V, energies, wavefunctions):
    '''
    Plots the energy levels and shifted wavefunctions.
    '''
    plt.figure(figsize=(8, 5))
    for i, (E, psi) in enumerate(zip(energies, wavefunctions)):
        plt.plot(x, psi + E, label=f"n={i}, E={E:.3f}")
    plt.title(f"Numeric Bound States")
    plt.xlabel("Position $x$")
    plt.ylabel("Wavefunction $\\psi(x)$")
    plt.ylim(0, np.max(wavefunctions) + np.max(energies))
    plt.legend(loc='upper right')
    plt.grid(True)
    plt.show()

def find_states_and_plot(potential, N_in=50, eps_in=0.1, lam_start=0.0):
    '''
    Finds the first four states and plots them.
    '''
    solutions = SchrodingerSolver(N=N_in, epsilon=eps_in, potential_func=potential)
    energies = solutions.find_eigenvalues(start_lambda=lam_start)
    eigenvectors = [solutions.compute_eigenvector(E, i) for i, E in enumerate(energies)]
    # plot_quantum_states(solutions.x, solutions.V, energies, eigenvectors)
    return solutions, energies, eigenvectors
```

```python
def anharmonic_potential(x, b=0.02):
    return 0.5 * x**2 + b * x**4 * (x**2 + 1) / 6

def first_order_energy(n, b=0.02):
    if n == 0:
        return 0.5 + 7 * b / 16
    if n == 1:
        return 1.5 + 45 * b / 16

def second_order_energy(n, b=0.02):
    if n == 0:
        return 0.5 + 7 * b / 16 - 567 * b**2 / 256
    if n == 1:
        return 1.5 + 45 * b / 16 - 7065 * b**2 / 256

b_range = np.linspace(0.001, 0.1, 100)
ground_energies = []
excited_energies = []
plt.figure(figsize=(8, 5))
for b in b_range:
    solutions, energies, wavefunctions = find_states_and_plot(lambda x: anharmonic_potential(x, b), N_in=500, eps_in=0.01, lam_start=0.0)
    ground_energies.append(energies[0])
    excited_energies.append(energies[1])

plt.plot(b_range, ground_energies, label='Ground State Energy')
plt.plot(b_range, excited_energies, label='Excited State Energy')
plt.plot(b_range, [first_order_energy(0, b) for b in b_range], '--', label='1st Order Ground Energy')
plt.plot(b_range, [first_order_energy(1, b) for b in b_range], '--', label='1st Order Excited Energy')
plt.plot(b_range, [second_order_energy(0, b) for b in b_range], '--', label='2nd Order Ground Energy')
plt.plot(b_range, [second_order_energy(1, b) for b in b_range], '--', label='2nd Order Excited Energy')
plt.xlabel('b')
plt.ylabel('Energy')
plt.title('Ground and Excited Energy vs. b')
plt.legend()
plt.grid(True)
plt.show()
```

For small $b$, perturbation theory works very well and both first and second order perturbations lie close to the numerical approximation. For larger $b$, the approximation begins to break down, particularly for the excited state. For the excited state, the wavefunction extends further out in $x$. Since the perturbation scales as $x^6$, the excited state incorporates the perturbation much more strongly than the ground state.

Looking at large $b$, for the ground state, the 2nd order correction clearly closer to the numerical result than the 1st order perturbation. However, for the excited state the 1st order approximation is actually closer to the numerical values than the 2nd order is. This is due to the large coefficient in the 2nd order perturbation. To fix this, we could use an asymptotic expansion for large $b$ and rely on the perturbation expansion for small $b$.
