Consider the double-well potential
\begin{equation}
    V(x) = \frac{1}{9d^4}(x^2 - d^2)^2\left(d^2 + \frac{x^2}{8}\right).
\end{equation}
This problem is not so easy to study by perturbation theory. Instead we use trial wavefunctions and the variational method.

We introduce the change of variable $y = x - d$ and substitute back
\begin{align}
    V
    &= \frac{1}{9d^4} (y^4 + 4dy^3 + 4d^2y^2) \left[\frac{1}{8}(9d^2 + 2dy + y^2)\right] \\
    &= \cdots = \frac{1}{2}y^2 + V_3 y^3 + \frac{7}{24d^2}y^4 + V_5 y^5 + \frac{1}{72d^4}y^6,
\end{align}
where we can identify the constants $V_3$ and $V_5$
\begin{equation}
    V_3 = \frac{11}{18}d^{-1}, \quad V_5 = \frac{1}{12}d^{-3}.
\end{equation}
Observe that for small $y$, the potential $V$ resembles the harmonic oscillator. A similar result occurs if we change variable to $y = x + d$ instead as the potential is symmetric in $x$. Based on this observation we introduce two wavefunctions that are ground states of the relevant harmonic oscillators
\begin{align}
    \psi^+(x) &= \frac{1}{\pi^{1/4}} e^{-\frac{1}{2}(x-d)^2}, \\
    \psi^-(x) &= \frac{1}{\pi^{1/4}} e^{-\frac{1}{2}(x+d)^2}.
\end{align}
Define two trial wavefunctions as
\begin{equation}
    \phi^\pm = C_\pm(\psi^+ \pm \psi^-).
\end{equation}

---

We require the trial wavefunctions to be normalised
\begin{equation}
    \langle \phi_{\pm} | \phi_{\pm} \rangle = 1.
\end{equation}
Expanding the inner product gives
\begin{equation}
    |C_{\pm}|^2 (\langle \psi_+ | \psi_+ \rangle \pm \langle \psi_+ | \psi_- \rangle \pm \langle \psi_- | \psi_+ \rangle + \langle \psi_- | \psi_- \rangle) = 1.
\end{equation}
The wavefunctions $\psi_+$ and $\psi_-$ are normalised ground states of harmonic oscillators centered at $+d$ and $-d$ respectively, so $\langle \psi_+ | \psi_+ \rangle = \langle \psi_- | \psi_- \rangle = 1$. The cross terms constitute the overlap integral,
\begin{align}
    \langle \psi_+ | \psi_- \rangle
    &= \int_{-\infty}^{\infty} \left(\frac{1}{\pi^{1/4}}e^{-\frac{1}{2}(x-d)^2}\right) \left(\frac{1}{\pi^{1/4}}e^{-\frac{1}{2}(x+d)^2}\right) \,dx \\
    &= \frac{1}{\sqrt{\pi}} \int_{-\infty}^{\infty} e^{-(x^2 + d^2)} \,dx \\
    &= \frac{e^{-d^2}}{\sqrt{\pi}} \int_{-\infty}^{\infty} e^{-x^2} \,dx = e^{-d^2}.
\end{align}
Substituting this back into the normalization equation yields
\begin{equation}
    |C_{\pm}|^2 (1 \pm e^{-d^2} \pm e^{-d^2} + 1) = 1.
\end{equation}
Thus, the normalisation constants are:
\begin{equation}
    C_{\pm} = \frac{1}{\sqrt{2(1 \pm e^{-d^2})}}.
\end{equation}

---

The energy expectation value is given by
\begin{equation}
    E_{\pm} = \langle \phi_{\pm} | H | \phi_{\pm} \rangle.
\end{equation}
Using the expansion of $\phi_{\pm}$ and noting that $H$ is symmetric, so $\langle \psi_+ | H | \psi_+ \rangle = \langle \psi_- | H | \psi_- \rangle$ and $\langle \psi_+ | H | \psi_- \rangle = \langle \psi_- | H | \psi_+ \rangle$, we get
\begin{equation}
    E_{\pm} = |C_{\pm}|^2 ( 2\langle \psi_+ | H | \psi_+ \rangle \pm 2\langle \psi_+ | H | \psi_- \rangle ).
\end{equation}
Substituting $|C_{\pm}|^2 = \frac{1}{2(1 \pm e^{-d^2})}$,
\begin{equation}
    E_{\pm} = \frac{\langle \psi_+ | H | \psi_+ \rangle \pm \langle \psi_+ | H | \psi_- \rangle}{1 \pm e^{-d^2}}.
\end{equation}
Compute
\begin{equation}
    A = \langle \psi_+ | H | \psi_+ \rangle = \langle \psi_+ | -\frac{1}{2}\frac{d^2}{dx^2} + V(x) | \psi_+ \rangle.
\end{equation}
Recall that $\psi_+$ is the ground state of a harmonic oscillator centered at $x=d$. Let $H_0^+ = -\frac{1}{2}\frac{d^2}{dx^2} + \frac{1}{2}(x-d)^2$, so $H_0^+ \psi_+ = \frac{1}{2} \psi_+$. We can rewrite the Hamiltonian $H$ as:
\begin{equation}
    H = H_0^+ - \frac{1}{2}(x-d)^2 + V(x).
\end{equation}
Substituting this into the expectation value gives
\begin{align}
    A
    &= \langle \psi_+ | H_0^+ | \psi_+ \rangle + \langle \psi_+ | V(x) - \frac{1}{2}(x-d)^2 | \psi_+ \rangle \\
    &= \frac{1}{2} + \langle \psi_+ | V(x) - \frac{1}{2}(x-d)^2 | \psi_+ \rangle.
\end{align}
Using the formula for $V$ calculated before, we calculate its expectation value in the state $\psi_+(y) \propto e^{-y^2/2}$. The expectation values of odd powers are zero due to symmetry. For even powers, we use the standard Gaussian moments, $\langle y^4 \rangle = \frac{3}{4}$ and $\langle y^6 \rangle = \frac{15}{8}$,
\begin{align}
    A
    &= \frac{1}{2} + \frac{7}{24d^2}\left(\frac{3}{4}\right) + \frac{1}{72d^4}\left(\frac{15}{8}\right) \\
    &= \frac{1}{2} + \frac{7}{32d^2} + \frac{5}{192d^4}.
\end{align}

---

Now consider
\begin{equation}
    B = \langle \psi_+ | H | \psi_- \rangle = \int_{-\infty}^{\infty} \psi_+(x) \left[ -\frac{1}{2}\frac{d^2}{dx^2} + V(x) \right] \psi_-(x) \,dx.
\end{equation}
Using the Schrödinger equation for $\psi_-$, which corresponds to a harmonic oscillator centered at $-d$
\begin{equation}
    \left( -\frac{1}{2}\frac{d^2}{dx^2} + \frac{1}{2}(x+d)^2 \right) \psi_- = \frac{1}{2} \psi_- \implies -\frac{1}{2}\frac{d^2\psi_-}{dx^2} = \left[ \frac{1}{2} - \frac{1}{2}(x+d)^2 \right] \psi_-.
\end{equation}
Substitute this into the integral and use $\psi_+(x)\psi_-(x) = \frac{1}{\sqrt{\pi}} e^{-(x^2+d^2)}$,
\begin{align}
    B
    &= \int_{-\infty}^{\infty} \psi_+(x) \psi_-(x) \left[ \frac{1}{2} - \frac{1}{2}(x+d)^2 + V(x) \right] \,dx \\
    &= \frac{e^{-d^2}}{\sqrt{\pi}} \int_{-\infty}^{\infty} e^{-x^2} \left[ \frac{1}{2} - \frac{1}{2}(x+d)^2 + V(x) \right] \,dx.
\end{align}
Expanding $V(x)$ around $x=0$ and integrating term-by-term using $\frac{1}{\sqrt{\pi}}\int x^{2n}e^{-x^2}dx = (2n-1)!!/2^n$, yields
\begin{equation}
    B = e^{-d^2} \left( -\frac{7d^2}{18} + \frac{7}{48} + \frac{1}{16d^2} + \frac{5}{192d^4} \right).
\end{equation}
Substituting $A$ and $B$ back into the expression for $E_{\pm}$ give the result
\begin{equation}
    E_{\pm} = \frac{A \pm B}{1 \pm e^{-d^2}}.
\end{equation}

---

When $d \gg 1$, the two potential wells are separated by a large distance compared to the natural width of the wavefunction. Physically, this corresponds to two independent harmonic oscillators. The potential barrier between the two wells, centered at $x=0$, becomes very high and wide. The overlap integral between the wavefunctions in the left and right wells is proportional to $e^{-d^2}$. As $d$ becomes large, this overlap vanishes almost completely. This means the probability of the particle tunneling from one well to the other becomes negligible.

We can determine the behavior of the energy by examining the limits of the constants $A$ and $B$ as $d \to \infty$.
The limit of $A$, the self-energy, as $d \to \infty$ is $A \approx \frac{1}{2}$. The limit of $B$, the interaction energy, as $d \to \infty$ is $B \to 0$. Hence,
\begin{equation}
    E_+ \approx E_- \approx \frac{1}{2}.
\end{equation}
The energy splitting between the symmetric state $E_+$ and the antisymmetric state ($E_-$) disappears. The two states become degenerate. The energy converges to $1/2$, the ground state energy of a single, isolated harmonic oscillator.

```python
import numpy as np
import matplotlib.pyplot as plt

class SchrodingerSolver:
    '''
    A solver for the 1D Schrödinger equation using the matrix method and Given's proceedure.
    '''
    def __init__(self, N, epsilon, potential_func):
        '''
        Initialise the solver with grid parameters and a potential function.
        Args:
            N: Half-number of grid points (total points = 2N).
            epsilon: Grid spacing.
            potential_func: A function V(x) that returns the potential.
        '''
        self.N = N
        self.epsilon = epsilon
        self.num_points = 2 * N

        # Discretise space
        self.x = np.linspace(-N * epsilon, N * epsilon, self.num_points, endpoint=False)

        # Diagonal elements
        self.V = potential_func(self.x)
        self.c = 1.0 + (epsilon**2) * self.V

        # Off-diagonal elements
        self.b_squared = 0.25

    def _count_eigenvalues_below(self, lam):
        '''
        Uses Given's Procedure to count eigenvalues < lambda, i.e., s(lambda).
        '''
        count = 0
        q = self.c[0] - lam

        if q < 0:
            count += 1

        for n in range(1, self.num_points):
            # Handling small q to avoid division by zero
            if abs(q) < 1e-12:
                q = 1e-12
            q = (self.c[n] - lam) - self.b_squared / q

            if q < 0:
                count += 1
        return count

    def find_eigenvalues(self, start_lambda=0.0):
        '''
        Finds the first four eigenvalues starting search from start_lambda.
        Returns a list of Energies (E), not matrix eigenvalues (lambda).
        '''
        energies = []
        current_lambda = start_lambda
        step_size = 0.5 * (self.epsilon**2)

        while len(energies) < 4:
            s_curr = self._count_eigenvalues_below(current_lambda)
            next_lambda = current_lambda + step_size
            s_next = self._count_eigenvalues_below(next_lambda)

            # If count increases, then an eigenvalue is crossed
            if s_next > s_curr:
                # Jumped over a state
                if s_next - s_curr > 1:
                    step_size /= 2
                    continue # Retry this step with smaller size

                # Refine with bisection
                lam_refined = self._bisect(current_lambda, next_lambda, s_curr)

                # Convert matrix lambda to energy E
                E = lam_refined / (self.epsilon**2)
                energies.append(E)

                # Move search forward
                current_lambda = lam_refined + (1e-4 * self.epsilon**2)
            else:
                current_lambda = next_lambda

            # Safety break for infinite loops
            if current_lambda > 100:
                print("Search limit exceeded.")
                break

        return np.array(energies)

    def _bisect(self, lower, upper, target_s_count):
        '''
        Bisection method to refine eigenvalue location.
        '''
        while upper - lower > 1e-12:
            mid = (lower + upper) / 2.0
            if self._count_eigenvalues_below(mid) > target_s_count:
                upper = mid
            else:
                lower = mid
        return (lower + upper) / 2.0

    def compute_eigenvector(self, E, state_index):
        '''
        Computes the eigenvector using symmetry to avoid tail divergence.
        '''
        psi = np.zeros(self.num_points)
        lam = (self.epsilon**2) * E

        # Integrate only from left boundary (n=0) to center (n=N)
        psi[0] = 1.0
        psi[1] = 2 * (self.c[0] - lam) * psi[0]

        # Stop at the center point x=0
        for n in range(1, self.N):
            psi[n+1] = 2 * (self.c[n] - lam) * psi[n] - psi[n-1]

            # Rescale to prevent overflow during calculation
            if abs(psi[n+1]) > 1e10:
                psi[:n+2] /= np.max(np.abs(psi[:n+2]))

        # Mirror the result to the right side around the center index N
        parity = 1 if (state_index % 2 == 0) else -1
        for k in range(1, self.N):
            if self.N + k < self.num_points:
                psi[self.N + k] = parity * psi[self.N - k]

        # Normalise
        norm = np.sqrt(np.sum(psi**2) * self.epsilon)
        psi = psi / norm

        # Convention: The main lobe on the right side (x > 0) is positive.
        right_half = psi[self.N:]
        # If the main feature on the right is negative, flip the whole wavefunction
        max_val_right = right_half[np.argmax(np.abs(right_half))]
        if max_val_right < 0:
            psi *= -1.0

        return psi

def plot_quantum_states(x, V, energies, wavefunctions):
    '''
    Plots the energy levels and shifted wavefunctions.
    '''
    plt.figure(figsize=(8, 5))
    plt.plot(x, V, label='Potential')
    for i, (E, psi) in enumerate(zip(energies, wavefunctions)):
        plt.plot(x, psi + E, label=f"n={i}, E={E:.3f}")
    plt.title(f"Numeric Bound States")
    plt.xlabel("Position $x$")
    plt.ylabel("Wavefunction $\\psi(x)$")
    plt.ylim(0, np.max(wavefunctions) + np.max(energies))
    plt.legend(loc='upper right')
    plt.grid(True)
    plt.show()

def find_states_and_plot(potential, N_in=50, eps_in=0.1, lam_start=0.0, plot=False):
    '''
    Finds the first four states and plots them.
    '''
    solutions = SchrodingerSolver(N=N_in, epsilon=eps_in, potential_func=potential)
    energies = solutions.find_eigenvalues(start_lambda=lam_start)
    eigenvectors = [solutions.compute_eigenvector(E, i) for i, E in enumerate(energies)]
    if plot:
        plot_quantum_states(solutions.x, solutions.V, energies, eigenvectors)
    return solutions, energies, eigenvectors
```

```python
def double_well_potential(x, d):
    return (x**2 - d**2)**2 * (d**2 + x**2 / 8) / (9 * d**4)

def theoretical_energies(n, d):
    A = (1 / 2) + (7 / (32 * d**2)) + (5 / (192 * d**4))
    B = np.exp(-d**2) * (-(7 * d**2) / 18 + (7 / 48) + (1 / (16 * d**2)) + (5 / (192 * d**4)))
    if n == 0:
        return (A + B) / (1 + np.exp(-d**2))
    if n == 1:
        return (A - B) / (1 - np.exp(-d**2))

beta_values = np.linspace(0.001, 1.0, 100)
d_values = 1 / np.power(9 * beta_values, 1 / 4)
ground_energies = []
excited_energies = []

for d in d_values:
    solutions, energies, _ = find_states_and_plot(lambda x: double_well_potential(x, d), N_in=500, eps_in=0.01, lam_start=0.0)
    ground_energies.append(energies[0])
    excited_energies.append(energies[1])
theoretical_ground_energies = [theoretical_energies(0, d) for d in d_values]
theoretical_excited_energies = [theoretical_energies(1, d) for d in d_values]

plt.figure(figsize=(8, 5))
plt.plot(beta_values, ground_energies, '-', label='Ground State')
plt.plot(beta_values, excited_energies, '-', label='Excited State')
plt.plot(beta_values, theoretical_ground_energies, '--', label='Theoretical Ground State')
plt.plot(beta_values, theoretical_excited_energies, '--', label='Theoretical Excited State')
plt.gca().invert_xaxis()
plt.xlabel(r'$\beta$')
plt.ylabel(r'Energy')
plt.title(r'Energy vs. $\beta$')
plt.legend()
plt.grid(True)
```

The ground state eigenvector $\Psi_0$ has even parity. This is because the potential $V(x)$ is an even function of $x$. The ground state of a 1D quantum system has zero nodes. A real, continuous wavefunction with no nodes must have the same sign everywhere. To satisfy the symmetry condition $\Psi(-x) = \pm \Psi(x)$ without changing sign, it must be even $\Psi(-x) = +\Psi(x)$.

The first excited state eigenvector $\Psi_1$ has odd parity. The eigenstates of a 1D potential are non-degenerate and alternate in parity as the energy increases. The first excited state ($n=1$) has exactly one node. For a symmetric potential, this node must occur at the center of symmetry. Therefore, the function must change sign at the origin, implying $\Psi(-x) = -\Psi(x)$.

We wish to show that the variational energies $E_+$ and $E_-$ are upper bounds for the true ground state energy $E_0$ and the true first excited state energy $E_1$, respectively. Let the complete set of exact eigenstates of the Hamiltonian be denoted by $|n\rangle$ with energies $E_n$, where $E_0 < E_1 < E_2 < \cdots$. Then

*   The set of even eigenstates is $\{|0\rangle, |2\rangle, |4\rangle, \dots\}$.
*   The set of odd eigenstates is $\{|1\rangle, |3\rangle, |5\rangle, \dots\}$.

Since $\psi_+(x) = \psi_-(-x)$, we have
\begin{equation}
    \phi_+(-x) \propto \psi_+(-x) + \psi_-(-x) = \psi_-(x) + \psi_+(x) = \phi_+(x).
\end{equation}
Thus, $\phi_+$ is an even function. When we expand $|\phi_+\rangle$ in the basis of exact eigenstates, the coefficients for all odd states must be zero. Therefore, $|\phi_+\rangle$ involves only the even states,
\begin{equation}
    |\phi_+\rangle = \sum_{n \in \text{Even}} c_n |n\rangle = c_0|0\rangle + c_2|2\rangle + c_4|4\rangle + \dots,
\end{equation}
where the expectation value is
\begin{equation}
    E_+ = \langle \phi_+ | H | \phi_+ \rangle = \sum_{n \in \text{Even}} |c_n|^2 E_n.
\end{equation}
Since $E_0$ is the lowest energy in the set $\{E_0, E_2, E_4, \dots\}$, we have that $E_n \geq E_0$ for all terms in the sum, so
\begin{equation}
    E_+ \geq E_0 \sum_{n \in \text{Even}} |c_n|^2 = E_0.
\end{equation}
Hence, $E_+$ is an upper bound for the ground state energy $E_0$.

Similarly,
\begin{equation}
    \phi_-(-x) \propto \psi_+(-x) - \psi_-(-x) = \psi_-(x) - \psi_+(x) = -(\psi_+ - \psi_-) = -\phi_-(x).
\end{equation}
Thus, $\phi_-$ is an odd function. It is therefore orthogonal to all even eigenstates $\langle 0 | \phi_- \rangle = 0$. When we expand $|\phi_-\rangle$, it can only contain terms from the odd eigenstates,
\begin{equation}
    |\phi_-\rangle = \sum_{n \in \text{Odd}} d_n |n\rangle = d_1|1\rangle + d_3|3\rangle + d_5|5\rangle + \dots,
\end{equation}
where the expectation value is
\begin{equation}
    E_- = \langle \phi_- | H | \phi_- \rangle = \sum_{n \in \text{Odd}} |d_n|^2 E_n.
\end{equation}
The lowest energy in the set $\{E_1, E_3, E_5, \dots\}$ is $E_1$. Therefore, $E_n \geq E_1$ for all terms in this sum,
\begin{equation}
    E_- \geq E_1 \sum_{n \in \text{Odd}} |d_n|^2 = E_1.
\end{equation}
Hence, $E_-$ is an upper bound for the first excited state energy $E_1$.

---

The variational method always overestimates the energy because the trial wavefunction (a simple sum of Gaussians) is constrained and cannot relax perfectly into the true shape of the eigenstate.

For small $\beta$, the bounds are close** to the true energies. As the wells move apart, the anharmonic corrections become negligible near the well centers. The potential locally looks almost exactly like a harmonic oscillator. Since our trial wavefunctions are harmonic oscillator states, they become nearly perfect matches.

For large $\beta$, the bounds become much looser. As $d$ shrinks, the barrier between the wells lowers and the wells begin to merge. Here, the region between the wells is no longer well-approximated by the tails of two separate harmonic oscillators. The anharmonic terms become significant, distorting the wavefunction away from a pure Gaussian shape.

Typically, the variational approximation of the ground state $E_+$ is slightly more accurate than the excited state $E_-$. This is because the ground state is a simpler, nodeless structure.

For the double-well potential with $d \gg 1$, the lowest energy levels are deeply buried below the central peak. Using the potential formula,
\begin{equation}
    V(x) = \frac{1}{9d^4}(x^2 - d^2)^2 \left(d^2 + \frac{x^2}{8}\right),
\end{equation}
substitute $x=0$ to obtain the height, $V(0) = \frac{d^2}{9}$. We established that for large separation $d$, the ground state energy $E_0$ converges to the harmonic oscillator energy $E_0 \approx \frac{1}{2}$. Eventually, for large enough $d$, the group state energy $E_0$ is lower than the central peak.

For the energy level to be lower than the peak, we need $E_0 < V(0)$, i.e., $\ > \sqrt{9/2}$.

*   If $d \gtrsim \sqrt{9/2}$, then the ground state energy $E_0$ is lower than the central peak. This is the tunneling regime. Classically, a particle with this energy would be trapped in either the left well or the right well and could never cross the barrier. Quantum mechanically, the particle can tunnel through the classically forbidden region, leading to the energy splitting $E_+ \neq E_-$.
    
*   If $d \lesssim \sqrt{9/2}$, then the ground state energy $E_0$ is higher than the central peak. The barrier is a small bump at the bottom of the potential. The particle has enough energy to breach the top of the barrier. The wavefunction is delocalised across the entire space not because of quantum tunneling, but because it is classically allowed to be there.

*   For $d \gg 1$, the barrier height is $V(0) = 16/9$ but the ground energy is $E_0 \approx 1/2$. Here, the energy is well below the peak and the wells are treated as isolated oscillators.

---

Performing a numerical analysis of the splitting ratio,
\begin{equation}
    \text{Ratio} = \frac{\Delta E_{\text{Variational}}}{\Delta E_{\text{Numerical}}},
\end{equation}
we see that the trial wavefunction method gives a poor estimate of the energy splitting for large separation $d$. The variational method underestimates the splitting by a large factor, and this error grows exponentially as $d$ increases.

The trial wavefunctions $\phi_{\pm}$ are built from harmonic oscillator ground states: $\psi \sim e^{-(x-d)^2/2}$. This implies that the wavefunction decays into the barrier as $e^{-x^2/2}$. However, the true decay of the wavefunction inside a potential barrier $V(x)$ is governed by the WKB action
\begin{equation}
    \psi(x) \sim \exp\left( -\int \sqrt{2(V(x)-E)} \, dx \right).
\end{equation}
For the specific potential given, the barrier is lower and thinner than what a pure Gaussian decay assumes. The Gaussian tails decay too fast and the overlap integral $B$ is too small. Consequently, the predicted energy splitting $\Delta E_{\text{var}} \approx 2B$ is much smaller than the true splitting.

As separation increases $d \to \infty$ (or $\beta \to 0$), the tunneling probability vanishes $e^{-S_0} \to 0$, leading to the two wells becoming effectively isolated. The ground and excited energoes $E_\pm$ become degenerate and $\Delta E \to 0$. However, the relative error of the variational method worsens. The variational energy converges to zero too quickly, ensuring $\Delta E_{\text{var}} / \Delta E_{\text{true}} \to 0$ rather than $1$.
