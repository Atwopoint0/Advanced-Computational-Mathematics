Accretion discs are composed of fluid orbiting a central object. They evolve viscously so that matter falls inwards while angular momentum drifts outwards. Accretion discs are found in binary star systems, around forming stars and in active galactic nuclei. We use cylindrical polar coordinates $(R, \phi, z)$ and assume axisymmetry. When the central object dominates the gravitational field the angular velocity of the matter in the disc is Keplerian so that
\begin{equation}
\Omega = \left(\frac{GM}{R^3}\right)^{1/2},
\end{equation}
where $M$ is the mass of the central object and $G$ is the gravitational constant. The disc is made up of annuli of matter lying between $R$ and $R + \Delta R$ with mass $2\pi R\Delta R\Sigma$, where $\Sigma(R, t)$ is the surface density (with dimensions $ML^{-2}$) of the disc at time $t$.

The equation describing conservation of mass is
\begin{equation}
    R\frac{\partial\Sigma}{\partial t} + \frac{\partial}{\partial R}(R\Sigma V_R) = 0,
\end{equation}
where $V_R$ is the radial velocity in the disc. Conservation of angular momentum gives us
\begin{equation}
    R\frac{\partial(\Sigma R^2\Omega)}{\partial t} + \frac{\partial}{\partial R}(R\Sigma V_RR^2\Omega) = \frac{1}{2\pi}\frac{\partial \Gamma}{\partial R},
\end{equation}
where the viscous torque is
\begin{equation}
    \Gamma = 2\pi R\nu\Sigma R^2\Omega',
\end{equation}
where $\Omega' = \frac{d\Omega}{dR}$ and $\nu(R, \Sigma)$ is the viscosity in the disc.
Since $\Omega$ is independent of time, the term $R^2\Omega$ (specific angular momentum) depends only on $R$. We can pull it out of the time derivative in the first term
\begin{equation}
    R(R^2\Omega)\frac{\partial \Sigma}{\partial t}.
\end{equation}
For the second term (the advection term), we apply the product rule,
\begin{equation}
    \frac{\partial}{\partial R}(R\Sigma V_R R^2\Omega) = (R\Sigma V_R)\frac{\partial(R^2\Omega)}{\partial R} + (R^2\Omega)\frac{\partial(R\Sigma V_R)}{\partial R}
\end{equation}
Substituting these expansions back yields
\begin{equation}
    R(R^2\Omega)\frac{\partial \Sigma}{\partial t} + (R\Sigma V_R)\frac{\partial(R^2\Omega)}{\partial R} + (R^2\Omega)\frac{\partial(R\Sigma V_R)}{\partial R} = \frac{1}{2\pi} \frac{\partial \Gamma}{\partial R}.
\end{equation}
Now, using the conservation of mass
\begin{equation}
    \frac{\partial}{\partial R}(R\Sigma V_R) = -R\frac{\partial \Sigma}{\partial t},
\end{equation}
substituting and cancelling gives gives
\begin{equation}
    (R\Sigma V_R)\frac{\partial(R^2\Omega)}{\partial R} = \frac{1}{2\pi} \frac{\partial \Gamma}{\partial R}.
\end{equation}
Substitute the expression for $R\Sigma V_R$ back into mass conservation to obtain
\begin{equation}
    R\frac{\partial \Sigma}{\partial t} = - \frac{\partial}{\partial R} \left[ \frac{1}{2\pi (R^2\Omega)'} \frac{\partial \Gamma}{\partial R} \right].
\end{equation}


Recall $\Omega = (GM)^{1/2} R^{-3/2}$. The specific angular momentum $R^2\Omega$ is
\begin{equation}
    R^2\Omega = R^2 \cdot (GM)^{1/2}R^{-3/2} = (GM)^{1/2}R^{1/2}.
\end{equation}
The derivative with respect to $R$ is given by
\begin{equation}
    \frac{d(R^2\Omega)}{dR} = \frac{d}{dR}\left[ (GM)^{1/2}R^{1/2} \right] = \frac{1}{2}(GM)^{1/2}R^{-1/2}.
\end{equation}
From $\Gamma = 2\pi R \nu \Sigma R^2 \Omega'$, we can compute
\begin{equation}
    \Omega' = \frac{d}{dR}\left[ (GM)^{1/2}R^{-3/2} \right] = -\frac{3}{2}(GM)^{1/2}R^{-5/2}.
\end{equation}
Now substitute this into the $\Gamma$ equation
\begin{equation}
    \Gamma = 2\pi R \nu \Sigma \cdot R^2 \cdot \left( -\frac{3}{2}(GM)^{1/2}R^{-5/2} \right) = -3\pi (GM)^{1/2} \nu \Sigma R^{1/2}.
\end{equation}
We need to evaluate the term inside the derivative
\begin{align}
    \frac{1}{2\pi (R^2\Omega)'} \frac{\partial \Gamma}{\partial R}
    &= \left[ \frac{1}{\pi (GM)^{1/2}R^{-1/2}} \right] \left[ -3\pi (GM)^{1/2} \frac{\partial}{\partial R} (\nu \Sigma R^{1/2}) \right] \\
    &= -3 R^{1/2} \frac{\partial}{\partial R} (\nu \Sigma R^{1/2}) .
\end{align}
Substitute this back to finally obtain
\begin{align}
    R\frac{\partial \Sigma}{\partial t} &= - \frac{\partial}{\partial R} \left[ -3 R^{1/2} \frac{\partial}{\partial R} (\nu \Sigma R^{1/2}) \right] \\
    &= \frac{\partial}{\partial R} \left[ R^{1/2} \frac{\partial}{\partial R} (3 \nu \Sigma R^{1/2}) \right] \\
    &= \frac{1}{R} \frac{\partial}{\partial R} \left[ R^{1/2} \frac{\partial}{\partial R} (3 \nu \Sigma R^{1/2}) \right].
\end{align}
This is the basic equation that governs the evolution of the surface density of a Keplerian accretion disc. The viscosity may be a function of $\Sigma$, $R$ and $t$ and so this equation may be non-linear.

The mass accretion rate $\dot{m}$ through the disc is

\end{equation}

Recakk the expression for the mass flux
\begin{equation}
    R\Sigma V_R = \frac{1}{2\pi (R^2\Omega)'} \frac{\partial \Gamma}{\partial R}.
\end{equation}
For a Keplerian disc, $\Omega = \sqrt{GM}R^{-3/2}$, so $R^2\Omega = \sqrt{GM}R^{1/2}$. Taking the derivative with respect to $R$, we get
\begin{equation}
    (R^2\Omega)' = \frac{d}{dR}(\sqrt{GM}R^{1/2}) = \frac{1}{2}\sqrt{GM}R^{-1/2}.
\end{equation}
Also, $\Gamma = -3\pi \nu \Sigma \sqrt{GM} R^{1/2}$. Substitute these back into the mass flux equation,
\begin{align}
    R\Sigma V_R
    &= \frac{1}{2\pi \left( \frac{1}{2}\sqrt{GM}R^{-1/2} \right)} \frac{\partial}{\partial R} \left( -3\pi \nu \Sigma \sqrt{GM} R^{1/2} \right) \\
    &= \frac{-3\pi \sqrt{GM}}{\pi \sqrt{GM} R^{-1/2}} \frac{\partial}{\partial R} (\nu \Sigma R^{1/2}) \\
    &= -3R^{1/2} \frac{\partial}{\partial R} (\nu \Sigma R^{1/2}).
\end{align}
Divide both sides by $R\Sigma$,
\begin{equation}
    V_R = \frac{-3R^{1/2}}{R\Sigma} \frac{\partial}{\partial R} (\nu \Sigma R^{1/2}) = -\frac{3}{\Sigma R^{1/2}} \frac{\partial}{\partial R} (\nu \Sigma R^{1/2}).
\end{equation}

We want the analytical solution for $\nu\Sigma$ assuming steady state and constant accretion rate $\dot{m}$. In a steady state, $\frac{\partial \Sigma}{\partial t} = 0$. Then
\begin{equation}
    \frac{1}{R}\frac{\partial}{\partial R} \left[ R^{1/2} \frac{\partial}{\partial R} (3\nu\Sigma R^{1/2}) \right] = 0.
\end{equation}
Multiply by $R$ and integrate once
\begin{equation}
    R^{1/2} \frac{\partial}{\partial R} (3\nu\Sigma R^{1/2}) = C_1.
\end{equation}
where $C_1$ is an integration constant. The accretion rate is given by $\dot{m} = -2\pi R \Sigma V_R$. Substitute the expression for $V_R$
\begin{equation}
    \dot{m} = -2\pi R \Sigma \left[ -\frac{3}{\Sigma R^{1/2}} \frac{\partial}{\partial R} (\nu \Sigma R^{1/2}) \right] = 6\pi R^{1/2} \frac{\partial}{\partial R} (\nu \Sigma R^{1/2}).
\end{equation}
Now compare this to our integrated equation to obtain
\begin{equation}
    C_1 = 3 \left( \frac{\dot{m}}{6\pi} \right) = \frac{\dot{m}}{2\pi}
\end{equation}
Substitute $C_1$ back into the differential equation
\begin{equation}
    R^{1/2} \frac{\partial}{\partial R} (3\nu\Sigma R^{1/2}) = \frac{\dot{m}}{2\pi},
\end{equation}
and divide by $R^{1/2}$
\begin{equation}
    \frac{\partial}{\partial R} (3\nu\Sigma R^{1/2}) = \frac{\dot{m}}{2\pi} R^{-1/2}.
\end{equation}
Integrate with respect to $R$
\begin{align}
    3\nu\Sigma R^{1/2}
    &= \frac{\dot{m}}{2\pi} \int R^{-1/2} dR \\
    &= \frac{\dot{m}}{2\pi} (2R^{1/2}) + C_2 \\
    &= \frac{\dot{m}}{\pi} R^{1/2} + C_2,
\end{align}
Apply the boundary conditions $\Sigma = 0$ at $R = R_{\text{in}}$. Since $\nu$ is finite, $\nu\Sigma = 0$ at $R = R_{\text{in}}$. Substitute this into the equation to obtain
\begin{equation}
    3(0) R_{\text{in}}^{1/2} = \frac{\dot{m}}{\pi} R_{\text{in}}^{1/2} + C_2 \\
\end{equation}
which implies $C_2 = -\frac{\dot{m}}{\pi} R_{\text{in}}^{1/2}$. Then
\begin{equation}
        3\nu\Sigma R^{1/2} = \frac{\dot{m}}{\pi} R^{1/2} - \frac{\dot{m}}{\pi} R_{\text{in}}^{1/2},
\end{equation}
where we divide by $3R^{1/2}$ to finally get
\begin{equation}
    \nu\Sigma = \frac{\dot{m}}{3\pi} - \frac{\dot{m}}{3\pi} \frac{R_{\text{in}}^{1/2}}{R^{1/2}} = \frac{\dot{m}}{3\pi} \left[ 1 - \sqrt{\frac{R_{\text{in}}}{R}} \right].
\end{equation}

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(1, 100, 500)
y = (1 / (3 * np.pi)) * (1 - np.sqrt(1/x))

plt.figure(figsize=(8, 5))
plt.plot(x, y, label=r'$\frac{\nu \Sigma}{\dot{m}} = \frac{1}{3\pi}\left(1 - \sqrt{\frac{R_{in}}{R}}\right)$')
plt.xlabel(r'$R / R_{in}$')
plt.ylabel(r'$\nu \Sigma$ (in units of $\dot{m}$)')
plt.title(r'Steady State Viscous Surface Density Profile')
plt.grid(True)
plt.xlim(1, 100)
plt.ylim(0, 0.12)
plt.axhline(1/(3*np.pi), color='r', linestyle='--', label=r'Asymptote $\frac{1}{3\pi}$')
plt.legend()
plt.show()
```

Define the dimensionless variables $R = rR_0$, $t = \tau t_0$, $\Sigma = \sigma \Sigma_0$, $\nu = \eta \nu_0$ and substitute. The left hand side is
\begin{equation}
    \frac{\partial (\sigma \Sigma_0)}{\partial (\tau t_0)} = \frac{\Sigma_0}{t_0} \frac{\partial \sigma}{\partial \tau}.
\end{equation}
The right hand side is
\begin{equation}
    \frac{1}{rR_0} \frac{\partial}{\partial (rR_0)} \left[ (rR_0)^{1/2} \frac{\partial}{\partial (rR_0)} (3 (\eta \nu_0) (\sigma \Sigma_0) (rR_0)^{1/2}) \right] = \frac{\nu_0 \Sigma_0}{R_0^2} \left( \frac{1}{r} \frac{\partial}{\partial r} \left[ r^{1/2} \frac{\partial}{\partial r} (3 \eta \sigma r^{1/2}) \right] \right).
\end{equation}
Equating, we see that for the dimensionless equation to be identical to the original (i.e., for the pre-factors to vanish/cancel out), we must have
\begin{equation}
    \frac{\Sigma_0}{t_0} = \frac{\nu_0 \Sigma_0}{R_0^2}.
\end{equation}
This gives the viscous timescale condition $t_0 = \frac{R_0^2}{\nu_0}$.

From the previous analysis, the dimensionless form is
\begin{equation}
    \frac{\partial \sigma}{\partial \tau} = \frac{1}{r} \frac{\partial}{\partial r} \left[ r^{1/2} \frac{\partial}{\partial r} (3\eta\sigma r^{1/2}) \right],
\end{equation}
where $\sigma = \Sigma/\Sigma_0$, $\eta = \nu/\nu_0$, $r = R/R_0$, and $\tau = t/t_0$. Using the substitution $X = r^{1/2}$,
\begin{equation}
    \frac{dr}{dX} = 2X \implies \frac{\partial}{\partial r} = \frac{dX}{dr}\frac{\partial}{\partial X} = \frac{1}{2X}\frac{\partial}{\partial X}.
\end{equation}
Now we substitute these terms into the RHS of the evolution equation:
*   Let $H = 3\eta\sigma r^{1/2}$. In terms of $X$, this is
    \begin{equation}
        H = 3\eta\sigma X.
    \end{equation}
    
*   The derivative $\frac{\partial}{\partial r} (3\eta\sigma r^{1/2})$ becomes
    \begin{equation}
        \frac{1}{2X} \frac{\partial}{\partial X} (3\eta\sigma X).
    \end{equation}

*   Multiply by $r^{1/2} = X$,
    \begin{equation}
        r^{1/2} \left[ \frac{1}{2X} \frac{\partial H}{\partial X} \right] = X \cdot \frac{1}{2X} \frac{\partial H}{\partial X} = \frac{1}{2} \frac{\partial}{\partial X} (3\eta\sigma X).
    \end{equation}

*   Apply $\frac{1}{r} \frac{\partial}{\partial r}$ to the result and recall $\frac{1}{r} = \frac{1}{X^2}$ and $\frac{\partial}{\partial r} = \frac{1}{2X}\frac{\partial}{\partial X}$,
    \begin{equation}
        \frac{1}{X^2} \left( \frac{1}{2X} \frac{\partial}{\partial X} \right) \left[ \frac{1}{2} \frac{\partial}{\partial X} (3\eta\sigma X) \right]
        = \frac{1}{4X^3} \frac{\partial^2}{\partial X^2} (3\eta\sigma X).
    \end{equation}

Now equate this to the LHS
\begin{equation}
    \frac{\partial \sigma}{\partial \tau} = \frac{1}{4X^3} \frac{\partial^2}{\partial X^2} (3\eta\sigma X).
\end{equation}
Since $X$ is independent of time $\tau$,
\begin{equation}
    \frac{\partial}{\partial \tau} (4X^3 \sigma) = 4X^3 \frac{\partial \sigma}{\partial \tau} = \frac{\partial^2}{\partial X^2} (3\eta\sigma X).
\end{equation}
Comparing this to the form
\begin{equation}
    \frac{\partial f}{\partial \tau} = \frac{\partial^2 g}{\partial X^2}\
\end{equation}
we determine $f$ and $g$
\begin{equation}
        f = 4X^3 \sigma, \quad g = 3\eta\sigma X.
\end{equation}
