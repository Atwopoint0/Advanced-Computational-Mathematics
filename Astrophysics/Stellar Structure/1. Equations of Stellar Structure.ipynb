The physics of stars can be encapsulated in a set of differential equations which can be solved with appropriate boundary conditions. The most efficient method of solving the equations is by a relaxation technique. In order to converge, it relies on an initial guess to the solution that is not far from the actual solution. We aim to construct such an initial model of a star by a
shooting technique that directly integrates the equations from the boundaries, where conditions are varied, until the two solutions meet in the middle.

The structure of a spherically symmetric star of uniform and unchanging composition, in thermal equilibrium, can be described by four non-linear differential equations in five variables together with an equation of state and boundary conditions.

1.  Hydrostatic Equilibrium,
\begin{equation}
    \frac{dp}{dr} = -\rho \frac{Gm}{r^2},
\end{equation}
where $p$ is the pressure and $\rho$ the density at radius $r$, measured from the centre, $m$ is the mass interior to $r$ and $G = 6.6726 \times 10^{-11}\,\mathrm{m}^3\mathrm{kg}^{-1}\mathrm{s}^{-2}$ is Newton's gravitational constant.

2.  Mass continuity,
\begin{equation}
    \frac{dm}{dr} = 4\pi r^2\rho.
\end{equation}

3.  Energy generation,
\begin{equation}
    \frac{dL_r}{dr} = 4\pi r^2\rho\epsilon,
\end{equation}
where $L_r$ is the luminosity, the outward flow of energy through a sphere at radius $r$, and $\epsilon = \epsilon(\rho, T, \text{composition})$ is the energy generation rate per unit mass.

4.  Energy can be transported by radiation (or equivalently conduction) or by bulk convective motions. In the radiative case
\begin{equation}
    \frac{dT}{dr} = -\frac{3\kappa\rho L_r}{16\pi acr^2T^3},
\end{equation}
where $\kappa = \kappa(\rho, T, \text{composition})$ is the opacity, $a = 7.5646 \times 10^{-16}\,\mathrm{J\,m}^{-3}\mathrm{K}^{-4}$ is the radiation constant and $c = 2.9979 \times 10^8\,\mathrm{ms}^{-1}$ the speed of light in a vacuum.

The equation of state relates pressure to density, temperature and composition throughout the star $p = p(\rho, T, \text{composition})$. Appropriate boundary conditions at the centre are $m = 0$ and $L_r = 0$ at $r = 0$. At the surface, $r = R_*$, the radius of the photosphere, an Eddington approximation to a plane
parallel grey atmosphere leads to
\begin{equation}
    L_* = 4\pi R_*^2\sigma T^4,
\end{equation}
where $\sigma = ac/4$ is the Stefan–Boltzmann constant, $L_*$ is the bolometric luminosity of the star and
\begin{equation}
    p\kappa = \frac{2}{3}\frac{GM_*}{R_*^2},
\end{equation}
where $M_*$ is the stellar mass. There are four independent variables $p(r)$, $m(r)$, $L_r$ and $T(r)$, when $\rho(r)$ is determined, for which a unique solution can be found.


---

In practice it is better to use $m$ rather than $r$ as the independent variable and, for uniform composition, the solution is unique for a given stellar mass. We then apply the surface boundary conditions at $m = M_*$.

Rearranging the mass continuity equation gives the transformation for the derivative
\begin{equation}
    \frac{d}{dm} = \left( \frac{1}{4\pi r^2 \rho} \right) \frac{d}{dr}.
\end{equation}

1.  Derivative of $r^3$: Using the chain rule:
\begin{align}
    \frac{d(r^3)}{dm}
    &= \frac{d(r^3)}{dr} \frac{dr}{dm} \\
    &= 3r^2 \frac{1}{4\pi r^2 \rho} \\
    &= \frac{3}{4\pi \rho}.
\end{align}

2.  Derivative of $L_r$: Using the energy generation equation,
\begin{align}
    \frac{dL_r}{dm}
    &= \frac{dL_r}{dr} \frac{dr}{dm} \\
    &= (4\pi r^2 \rho \epsilon) \frac{1}{4\pi r^2 \rho} \\
    &= \epsilon.
\end{align}

3.  Derivative of $T^4$: First, differentiate $T^4$ with respect to $r$ using the chain rule and the radiative transport equation
\begin{align}
    \frac{d(T^4)}{dr}
    &= 4T^3 \frac{dT}{dr} \\
    &= 4T^3 \left( - \frac{3\kappa \rho L_r}{16\pi a c r^2 T^3} \right) \\
    &= - \frac{3\kappa \rho L_r}{4\pi a c r^2}
\end{align}
Now, convert to derivative with respect to $m$,
\begin{align}
    \frac{d(T^4)}{dm}
    &= \frac{d(T^4)}{dr} \frac{dr}{dm} \\
    &= \left( - \frac{3\kappa \rho L_r}{4\pi a c r^2} \right) \frac{1}{4\pi r^2 \rho} \\
    &= - \frac{3\kappa L_r}{16\pi^2 a c r^4}.
\end{align}

4.  Derivative of $\log p$: Using the chain rule and the hydrostatic equilibrium equation,
\begin{align}
    \frac{dp}{dm}
    &= \frac{dp}{dr} \cdot \frac{dr}{dm} \\
    &= \left( -\rho \frac{Gm}{r^2} \right) \cdot \frac{1}{4\pi r^2 \rho} \\
    &= - \frac{Gm}{4\pi r^4}.
\end{align}
Now divide by $p$ for the logarithm derivative
\begin{equation}
    \frac{d(\log p)}{dm} = - \frac{Gm}{4\pi r^4 p}.
\end{equation}

As they stand the equations are not suitable for numerical integration at the centre. It is therefore necessary to develop them to obtain conditions at some small but finite value of $r$.

At the center of the star, we assume density $\rho$, energy generation $\epsilon$, and opacity $\kappa$ tend towards constant central values $\rho_c$, $\epsilon_c$, and $\kappa_c$, respectively.

1.  Mass and radius ($r^3$): We have already established
\begin{equation}
    \frac{d(r^3)}{dm} = \frac{3}{4\pi \rho}.
\end{equation}
Near the center, $\rho \approx \rho_c$. We integrate with respect to $m$
\begin{equation}
    \int_0^{r^3} d(r'^3) = \int_0^m \frac{3}{4\pi \rho_c} \,dm',
\end{equation}
which evaluates to
$r^3 = \frac{3m}{4\pi \rho_c}$.

2.  Luminosity $L_r$: Again, we have
\begin{equation}
    \frac{dL_r}{dm} = \epsilon.
\end{equation}
Near the center, $\epsilon \approx \epsilon_c$. Integrating from the center where $m=0, L_r=0$,
\begin{equation}
    L_r = \epsilon_c m = \frac{4}{3}\pi \rho_c \epsilon_c r^3.
\end{equation}

3.  Pressure $p$: We can use the hydrostatic equilibrium equation combined with the mass relation. From basic structure equations
\begin{equation}
    \frac{dp}{dr} = -\frac{Gm\rho}{r^2}.
\end{equation}
Substitute the approximations $m \approx \frac{4}{3}\pi \rho_c r^3$ and $\rho \approx \rho_c$ to get
\begin{equation}
    \frac{dp}{dr} \approx -\frac{4}{3}\pi G \rho_c^2 r.
\end{equation}
Now, integrate from the center where $r=0 and p=p_c$, to a small radius $r$,
\begin{equation}
    \int_{p_c}^p dp' = -\frac{4}{3}\pi G \rho_c^2 \int_0^r r' \,dr'.
\end{equation}
This evaluates to
\begin{equation}
    p = p_c - \frac{2}{3}\pi G \rho_c^2 r^2.
\end{equation}

4.  Temperature $T^4$:
Recall that the derivative for $T^4$ with respect to $r$ was
\begin{equation}
    \frac{d(T^4)}{dr} = -\frac{3\kappa \rho L_r}{4\pi a c r^2}.
\end{equation}
Substitute the approximations $L_r \approx \frac{4}{3}\pi \rho_c \epsilon_c r^3$, $\rho \approx \rho_c$, and $\kappa \approx \kappa_c$,
\begin{equation}
    \frac{d(T^4)}{dr} \approx -\frac{\kappa_c \epsilon_c \rho_c^2}{a c} r.
\end{equation}
Integrate from the center where $r=0$ and $T=T_c$,
\begin{equation}
    T^4 - T_c^4 = -\frac{\kappa_c \epsilon_c \rho_c^2}{a c} \int_0^r r' \,dr',
\end{equation}
which yields
\begin{equation}
    T^4 = T_c^4 - \frac{\kappa_c \epsilon_c \rho_c^2}{2 a c} r^2.
\end{equation}

We shall assume that stars are composed entirely of hydrogen (mass fraction $X$, assumed to be $0.7$) and helium (mass fraction $Y = 1 - X$) and that the contributions to pressure, other than that of the perfect gas, are negligible
so that
\begin{equation}
    p = \frac{\rho R^*T}{\mu},
\end{equation}
where $R^* = 8.3145 \times 10^3\,\mathrm{J}\,\mathrm{kg}^{-1}\mathrm{K}^{-1}$ is the gas constant and $\mu$ is the mean molecular weight. It is sufficient to calculate $\mu$ on the assumption that the material is completely ionised so that
\begin{equation}
    \frac{1}{\mu} = 2X + \frac{3}{4}Y.
\end{equation}
The same applies to $\gamma$, the adiabatic exponent, which may be taken to be
constant at $5/3$ for a monatomic ideal gas. We shall further assume that the opacity $\kappa$, may be approximated by the contribution from electron scattering
\begin{equation}
    κ_{\text{es}} = 0.02(1 + X)\,\mathrm{m}^2\mathrm{kg}^{-1}.
\end{equation}
Finally, we assume that nuclear burning proceeds via a combination of the proton-proton chain and CNO cycle, producing energy at a rate
\begin{equation}
    \epsilon = \left(0.25 X^2 e^{-33.8T_6^{-1/3}} + 8.8 \times 10^{18} X e^{-152.28T_6^{-1/3}}\right)T_6^{-2/3} \frac{\rho}{\mathrm{kg\,m}^{-3}}\,\mathrm{W\,kg}^{-1}.
\end{equation}
where $T_6 = T/10^6 \,\mathrm{K}$.

The Sun is observed to have mass $M_\odot = 1.9891 \times 10^{30} \,\mathrm{kg}$, radius $R_\odot = 6.9598 \times 10^8 \,\mathrm{m}$ and luminosity $L_\odot = 3.8515 \times 10^{26}\,\mathrm{W}$.

To estimate the temperature required to halt gravitational collapse, we balance the thermal energy of the gas against the gravitational potential energy.

First, we must determine $\mu$ using the composition $X=0.7, Y=0.3$,
\begin{equation}
    \frac{1}{\mu} = 2X + \frac{3}{4}Y = \frac{1}{1.625} \approx 0.6154.
\end{equation}
For a star to be in hydrostatic equilibrium, the thermal energy per unit mass must be of the same order as the gravitational potential energy per unit mass,
\begin{equation}
    \frac{k_B T}{\mu m_H} \approx \frac{G M}{R}.
\end{equation}
The specific gas constant $R^*$ suggests the thermal energy scale is proportional to $R^* T / \mu$
\begin{equation}
    \frac{R^* T_c}{\mu} \approx \frac{G M_\odot}{R_\odot} \implies T_c \approx \frac{\mu G M_\odot}{R^* R_\odot} \approx 1.41 \times 10^7 \,\mathrm{K}.
\end{equation}
This temperature provides sufficient thermal pressure to balance the gravitational forces of a solar-mass object, satisfying the condition to halt gravitational collapse. Looking at the equation for energy generation $\epsilon$, the proton-proton chain term contains the factor $e^{-33.8 T_6^{-1/3}}$. At $T \approx 1.4 \times 10^7$ K (where $T_6 = 14$), this exponential factor becomes large enough for nuclear fusion to generate significant energy.

We approximate the pressure gradient $dp/dr$ as a constant slope from the center to the surface. Set the boundary conditions: at $r=0$ we have $p = p_c$ and at $r=R_\odot$, we have $p \approx 0$ (as surface pressure is negligible compared to the core). Then
\begin{equation}
    \left| \frac{dp}{dr} \right| \approx \frac{p_c - 0}{R_\odot} = \frac{p_c}{R_\odot}.
\end{equation}
To solve the hydrostatic equilibrium for an order-of-magnitude estimate, we substitute average values for the variables on the right-hand side, i.e., average density $\bar{\rho}$ and surface gravity scale $\frac{GM_\odot}{R_\odot^2}$. This gives
\begin{equation}
    \frac{p_c}{R_\odot} \approx \bar{\rho} \frac{G M_\odot}{R_\odot^2} \implies p_c \approx \bar{\rho} \frac{G M_\odot}{R_\odot}.
\end{equation}
The average density is approximately
\begin{equation}
    \bar{\rho} = \frac{M_\odot}{\frac{4}{3}\pi R_\odot^3} \approx 1408 \, \mathrm{kg m}^{-3}.
\end{equation}
The pressure is approximately
\begin{equation}
    p_c \approx 2.7 \times 10^{14} \, \mathrm{Pa}.
\end{equation}
