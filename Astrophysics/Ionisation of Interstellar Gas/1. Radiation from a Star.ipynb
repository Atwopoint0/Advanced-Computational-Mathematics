The interstellar medium surrounding a hot star is ionised by the radiation from the star. We aim to calculate the size of the ionised region, and gain some insight into how its structure depends on the nature of the radiation from the star.

Assuming that there is a uniform, static, constant temperature gas surrounding a spherically symmetric star provides a good approximation. Each gas element is assumed to be in ionisation equilibrium, so the ionisation rate for each atom is balanced by recombinations. We assume that the sole source of ionisation is by absorption of radiation from the star giving a bound electron enough energy to escape from the atom. Recombination occurs when a free electron is captured by an ion with the creation of a photon. Since hydrogen is the most common element in the Universe, its behaviour will dominate in most cases, so we consider only a pure hydrogen interstellar medium.

The radiation from the star is specified as $L_\nu$, the total energy output from the star per unit frequency $\nu$ per unit time, so the luminosity per unit frequency interval $L_\nu$ is expressed in $\mathrm{W Hz}^{-1}$. The total energy output radiated from the star is the integral of this quantity over all frequencies, or
\begin{equation}
    L = \int_0\infty L_\nu\,d\nu.
\end{equation}
In some cases a star spectrum is reasonably well approximated by a black-body, so the radiation flux in a frequency interval $d\nu$ at frequency $\nu$ emerging per unit area from the surface of the star where the temperature is $T_*$ is given by
\begin{equation}
    I_\nu \,d\nu = \frac{2\pi\hbar}{c^2} \frac{\nu^3}{\exp(\frac{\hbar\nu}{kT_âˆ—}) - 1}
\,d\nu,
\end{equation}
where $\hbar$ is Planck's constant, $c$ is the speed of light and $k$ is Boltzmann's constant. Thus for a star of radius $R$
\begin{equation}
    L_\nu = 4\pi R^2 \frac{2\pi\hbar}{c^2}\frac{\nu^3}{\exp(\frac{\hbar\nu}{kT+*}) - 1}.
\end{equation}
Substituting the equation for $L_\nu$ into the integral, we get
\begin{equation}
    L = 4\pi R^2 \int_{0}^{\infty} \frac{2\pi \hbar}{c^2} \frac{\nu^3}{e^{\frac{\hbar\nu}{kT_*}} - 1} \, d\nu.
\end{equation}
Solving this integral yields the Stefan-Boltzmann law,
\begin{equation}
    L = 4\pi R^2 \sigma T_*^4,
\end{equation}
where $\sigma \approx 5.67 \times 10^{-8} \, \mathrm{W m}^{-2} \mathrm{K}^{-4}$ is the Stefan-Boltzmann constant.

---

The sun has radius $R = 6.96 \times 10^{8}\,\mathrm{m}$, and total luminosity $L = 3.90 \times 10^{26} \,\mathrm{W}$. Rearrange the Stefan-Boltzmann law to solve for temperature,
\begin{equation}
    T_* = \left( \frac{L}{4\pi R^2 \sigma} \right)^{1/4}.
\end{equation}
Calculate the surface area:
\begin{equation}
    4\pi R^2 \approx 4\pi (6.96 \times 10^8)^2 \approx 6.087 \times 10^{18}\,\mathrm{m}^2.
\end{equation}
Then,
\begin{equation}
    T_* = (1.130 \times 10^{15})^{1/4} \approx 5796\,\mathrm{K} \approx 5800 \,\mathrm{K}.
\end{equation}

A $7$ solar mass star has a surface temperature of $T_* = 20000\,\mathrm{K}$ and a luminosity of $L = 4.0 \times 10^{29}\,\mathrm{W}$. Rearrange the Stefan-Boltzmann law to solve for the radius
\begin{equation}
    R = \sqrt{\frac{L}{4\pi \sigma T_*^4}}.
\end{equation}
This yield that the radius is approximately $R \approx 1.87 \times 10^9\,\mathrm{m}$.

A $12$ solar mass star has a surface temperature of $T_* = 25000\,\mathrm{K}$ and a luminosity of $L = 4.0 \times 10^{30}\,\mathrm{W}$. Using the same formula, the radius is approximately $R \approx 3.79 \times 10^9\,\mathrm{m}$.

An element of gas at a distance $r$ from the (centre of the) star will receive
\begin{equation}
    \frac{L_\nu}{4\pi r^2}\,\mathrm{Wm}^{-2}\mathrm{Hz}^{-1}
\end{equation}
if the radiation is not attenuated by any material between it and the star. The number of photons received per second per unit frequency interval is $L_\nu$ divided by the energy $h\nu$ per photon. If these photons have frequency $\nu \geq \nu_0 = 3.29 \times 10^{15}\,\mathrm{Hz}$, then they have enough energy to separate a hydrogen atom into a proton and an electron. The rate at which this happens depends on the frequency of the radiation, and is given by the photon rate per second multiplied by an absorption coefficient $a_\nu$ per hydrogen atom. For an energy flux $I_\nu(r)$, so photon flux $I_\nu(r)/\hbar\nu$, the rate of
ionisation per unit volume is
\begin{equation}
    n_{H^0} \int_{\nu_0}^\infty \frac{I_\nu(r)}{\hbar\nu} a_\nu\,d\nu,
\end{equation}
where $n_{H^0}$ is the number density of neutral hydrogen atoms (i.e. number of neutral hydrogen atoms per cubic metre). The coefficient $a_\nu$ depends only on the atomic species being considered, and for neutral hydrogen
\begin{align}
    a_\nu &= a_{\nu_0}\left(\frac{\nu_0}{\nu}\right)^3, &\nu \geq \nu_0, \\
    a_\nu &= 0, &\nu < \nu_0,
\end{align}
where $a_{\nu_0} = 6.3 \times 10^{-22}\,\mathrm{m}^2$. If absorption can occur at the gas element, then it can also occur in all the elements between the star and the one under consideration. As we go along a path $dr$ the radiation is attenuated, so at a given frequency $\nu$,
\begin{equation}
    \frac{dI_\nu}{dr} = -n_{H^0} a_\nu I_\nu,
\end{equation}
and so
\begin{equation}
    I_nu(r) = I_\nu(R)e^{-\tau_\nu},
\end{equation}
where $\tau_\nu$ is defined by
\begin{equation}
    \frac{d\tau_\nu}{dr} = n_{H^0}(r)a_\nu
\end{equation}
with $\tau_\nu(R) = 0$ and $\tau_\nu$ is referred to as the optical depth at the frequency $\nu$. There is no redistribution of the photons in frequency (they are effectively absorbed from the point of view of this calculation), so we can determine $\tau_\nu$ from
\begin{equation}
    \frac{d\tau_{\nu_0}}{dr} = n_{H^0}(r)a_{\nu_0}
\end{equation}
and
\begin{equation}
    \tau_\nu = \tau_{\nu_0}\left(\frac{\nu_0}{\nu}\right)^3.
\end{equation}
This absorption of radiation and the ionisation it causes must be balanced by recombination of protons and electrons at the same rate per unit volume. This rate depends on the number densities of the protons $n_p$ and the electrons $n_e$, their relative velocity and a velocity-dependent cross-section for the interaction which has to be integrated over the velocity distribution at
whatever temperature the gas is at. These velocity-dependent terms are combined into recombination coefficients $\alpha_B$ which are tabulated for various gas temperatures $T$:
\begin{array}{ | c | c | }
    \hline
    T\;(K) & \alpha_B\;(\mathrm{m}^3\mathrm{s}^{-1}) \\
    \hline
    5 000  & 4.54 \times 10^{-19} \\
    10 000 & 2.59 \times 10^{-19} \\
    20 000 & 2.52 \times 10^{-19} \\
    \hline
\end{array}
Then we are in a position to write down the ionisation balance equation
\begin{equation}
    n_{H^0} \int_{\nu_0}^\infty \frac{L_\nu}{4\pi r^2\hbar\nu} a_\nu e^{-\tau_\nu}\,d\nu = n_pn_e\alpha_B(T),
\end{equation}
and subsidiary equations
\begin{align}
    &\frac{d\tau_{\nu_0}}{dr} = n_{H^0}(r)a_{\nu_0}, \\
    &\tau_nu = \tau_{\nu_0}\left(\frac{\nu_0}{\nu}\right)^3, \\
    &n_p = n_e, \\
    &n_p + n_{H^0} = n_H,
\end{align}
which govern the ionisation balance for the hydrogen-filled interstellar medium near a star.

We have and assumed that the energy released by recombination does not give rise to an ionising photon. Often it does, but we assume that all this does is cause another ionisation nearby until recombination occurs to an upper level in the hydrogen atom, and then the energy is lost from the system through radiation at frequencies less than $\nu_0$. The net result is a change in the effective recombination coefficient, to the one quoted here.
