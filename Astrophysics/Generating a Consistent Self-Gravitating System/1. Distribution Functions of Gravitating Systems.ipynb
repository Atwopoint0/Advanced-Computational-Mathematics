A collection of $N$ particles moving under their mutual gravitational attraction only, can be well described by a continuous distribution function, $f(x, v, t)\,d^3x\,d^3v$, in the limits where $N$ is large and the particles have some positions $x_i$ and velocities $v_i$, $i \in \{1, \dots, N\}$, at time $t$.

In the limit where two-body interactions can be neglected ($N$ sufficiently large), the flow in phase space is said to be collisionless and the distribution function satisfies the Boltzmann equation
\begin{equation}
    \frac{df}{dt} = 0.
\end{equation}
The dynamics of the system are governed by the Poisson equation,
\begin{equation}
    \nabla^2\Phi(x) = 4\pi G\rho(x),
\end{equation}
where the spatial density is $\rho(x, t) = \int f\,d^3v$, $\Phi$ is the potential and $G$ is the gravitational constant.

We want to generate a discrete particle realisation of a spherical self-gravitating system, given a potential-density pair describing the system, and compare the results for a finite number of particles with those from continuous distributions. Here we will consider a specific example of a spherical, isotropic distribution, the so-called $\gamma$ distribution functions. The $\gamma$ functions have density profiles that are proportional to $r^{-\gamma}$ as $r \to 0$. Here we consider $\gamma = 0$.

For a spherical, isotropic system, $f$ can be expressed as a function of the specific energy $E$ only. For a particle $i$,
\begin{equation}
    E_i = \frac{1}{2} \nu_i^2 + \Phi(r_i),
\end{equation} where $r_i = |x_i|$ and $\nu_i = |v_i|$. The $\gamma = 0$ model is specified by the density profile
\begin{equation}
    \rho(r) = \frac{3Ma}{4\pi(r + a)^4}.
\end{equation}
For a spherical density distribution $\rho(r)$, the mass enclosed within a radius $r$ is given by the integral
\begin{equation}
    M(r) = \int_{0}^{r} 4\pi (r')^2 \rho(r') \, dr'.
\end{equation}
Substitute the given density profile to get
\begin{equation}
    M(r) = \int_{0}^{r} 4\pi (r')^2 \left( \frac{3Ma}{4\pi(r'+a)^4} \right) \, dr' = 3Ma \int_{0}^{r} \frac{(r')^2}{(r'+a)^4} \, dr'.
\end{equation}
To solve this integral, we use the substitution $u = r' + a$. Now the enclosed mass can be computed as follows,
\begin{align}
    M(r)
    &= 3Ma \int_{a}^{r+a} \frac{(u-a)^2}{u^4} \,du \\
    &= 3Ma \int_{a}^{r+a} (u^{-2} - 2au^{-3} + a^2 u^{-4}) \,du \\    
    &= 3Ma \left[ -\frac{1}{u} + \frac{a}{u^2} - \frac{a^2}{3u^3} \right]_{a}^{r+a} \\
    &= M \left[ 1 - \frac{3a}{r+a} + \frac{3a^2}{(r+a)^2} - \frac{a^3}{(r+a)^3} \right] \\
    &= M \left( 1 - \frac{a}{r+a} \right)^3 = M \left( \frac{r}{r+a} \right)^3.
\end{align}
The gradient of the potential for a spherical system is related to the enclosed mass by Newton's shell theorem
\begin{equation}
    \frac{d\Phi}{dr} = \frac{G M(r)}{r^2}.
\end{equation}
Substitute our expression for $M(r)$ to obtain
\begin{equation}
    \frac{d\Phi}{dr} = \frac{G}{r^2} \left( \frac{M r^3}{(r+a)^3} \right) = \frac{GM r}{(r+a)^3}.
\end{equation}
Now, integrate from infinity (where $\Phi = 0$) to $r$,
\begin{equation}
    \Phi(r) = \int_{\infty}^{r} \frac{GM r'}{(r'+a)^3} \, dr',
\end{equation}
and once again, use the substitution $u = r' + a$,
\begin{align}
    \int \frac{r'}{(r'+a)^3} \,dr'
    &= \int (u^{-2} - a u^{-3}) \,du \\
    &= -\frac{1}{u} + \frac{a}{2u^2} \\
    &= -\frac{1}{r'+a} + \frac{a}{2(r'+a)^2}.
\end{align}
Applying the limits,
\begin{align}
    \Phi(r)
    &= GM \left[ -\frac{1}{r+a} + \frac{a}{2(r+a)^2} \right] \\
    &= -\frac{GM(2r+a)}{2(r+a)^2} \\
    &= -\frac{GM}{2a}\left[\frac{(r+a)^2 - r^2}{(r+a)^2}\right].
\end{align}

Recall that the enclosed mass within radius $r$ is given by
\begin{equation}
    M(r) = \int_0^r 4\pi\rho(r')(r')^2\,dr' = M\left(\frac{r}{r + a}\right)^3.
\end{equation}
Without loss of generality, we take $a = 1$, $G = 1$, and $M(r \to \infty) = 1$. Additionally define the dimensionless binding energy $\epsilon = -E \times (a/GM)$ and potential $\Psi = -\Phi \times (a/GM)$. Given $\rho(\Phi)$, it is straightforward to derive $f(\epsilon)$ with an Abel transform. For this model
\begin{equation}
    f(\epsilon) = \frac{3}{4\sqrt{2}\pi^3} \int_0^\epsilon \frac{(1 - y)^2(2y + 4y^2)}{y^4\sqrt{\epsilon - \Psi}} \,d\Psi,
\end{equation}
where $y(\Psi) = \sqrt{1 - 2\Psi}$.

We switch the integration variable $y = \sqrt{1-2\Psi}$,
\begin{align}
    f(\epsilon)
    &= \frac{3}{4\sqrt{2}\pi^3} \int_{1}^{y_0} \frac{(1-y)^2(2y+4y^2)}{y^4 \left(\frac{1}{\sqrt{2}}\sqrt{y^2 - y_0^2}\right)} (-y) \,dy \\
    &= \frac{3}{4\pi^3} \int_{y_0}^{1} \frac{y (1-y)^2 (2y+4y^2)}{y^4 \sqrt{y^2 - y_0^2}} \,dy \\
    &= \frac{3}{4\pi^3} \int_{y_0}^{1} \frac{\frac{2}{y^2} - 6 + 4y}{\sqrt{y^2 - y_0^2}} \,dy \\
    &= \frac{3}{4\pi^3} \left[ 2\int_{y_0}^{1} \frac{dy}{y^2\sqrt{y^2 - y_0^2}} - 6\int_{y_0}^{1} \frac{dy}{\sqrt{y^2 - y_0^2}} + 4\int_{y_0}^{1} \frac{y dy}{\sqrt{y^2 - y_0^2}} \right].
\end{align}
We evaluate the three integrals separately:
*   Integral 1:    
\begin{align}
    \int_{y_0}^1 \frac{dy}{y^2\sqrt{y^2 - y_0^2}}
    &= \left[ \frac{\sqrt{y^2 - y_0^2}}{y_0^2 y} \right]_{y_0}^{1} \\
    &= \frac{\sqrt{1 - y_0^2}}{y_0^2} - 0 \\
    &= \frac{\sqrt{2\epsilon}}{1-2\epsilon}.
\end{align}

*   Integral 2:
\begin{align}
    \int_{y_0}^1 \frac{dy}{\sqrt{y^2 - y_0^2}}
    &= \left[ \cosh^{-1}\left(\frac{y}{y_0}\right) \right]_{y_0}^{1} \\
    &= \cosh^{-1}\left(\frac{1}{y_0}\right) - \cosh^{-1}(1) \\
    &= \sinh^{-1}\sqrt{\frac{2\epsilon}{1-2\epsilon}}.
\end{align}

*   Integral 3:
\begin{align}
    \int_{y_0}^1 \frac{y dy}{\sqrt{y^2 - y_0^2}}
    &= \left[ \sqrt{y^2 - y_0^2} \right]_{y_0}^{1} \\
    &= \sqrt{1 - y_0^2} = \sqrt{2\epsilon}.
\end{align}

Substitute the evaluated parts back into the expression for $f(\epsilon)$
\begin{align}
    f(\epsilon)
    &= \frac{3}{4\pi^3} \left[ 2\left(\frac{\sqrt{2\epsilon}}{1-2\epsilon}\right) - 6\left(\sinh^{-1}\sqrt{\frac{2\epsilon}{1-2\epsilon}}\right) + 4(\sqrt{2\epsilon}) \right] \\
    &= \frac{3}{2\pi^3} \left[ \left( \frac{\sqrt{2\epsilon}}{1-2\epsilon} + 2\sqrt{2\epsilon} \right) - 3\sinh^{-1}\sqrt{\frac{2\epsilon}{1-2\epsilon}} \right] \\
    &= \frac{3}{2\pi^3} \left[ \frac{(3-4\epsilon)\sqrt{2\epsilon}}{1-2\epsilon} - 3 \sinh^{-1} \sqrt{\frac{2\epsilon}{1-2\epsilon}} \right].
\end{align}

The incremental mass of particles, $dM$, with binding energies in the energy interval $\epsilon$ to $\epsilon + d\epsilon$ is given by the differential energy distribution
\begin{equation}
    \frac{dM}{d\epsilon} = f(\epsilon)g(\epsilon),
\end{equation}
where the density of states $g(\epsilon)$ is given by
\begin{equation}
    g(\epsilon) = 8\pi^2 \left[\sqrt{1 - 2\epsilon} \frac{3 - 14\epsilon - 8\epsilon^2}{12\epsilon^2} - \pi + \frac{1 - 6\epsilon + 16\epsilon^2}{(2\epsilon)^{5/2}}\cos^{-1}(-\sqrt{1 - 2\epsilon}) \right].
\end{equation}
We wish to generate a realisation of our $\gamma = 0$ distribution, using a Monte-Carlo acceptance-rejection algorithm.
