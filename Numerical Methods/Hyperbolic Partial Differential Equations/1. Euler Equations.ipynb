Conider the kinematic wave equation or inviscid Burger's equation
\begin{equation}
    u_t + f(u)_x = 0, \quad f(u) = \frac{1}{2}u^2.
\end{equation}
This posesses a non-linear flux term proportional to the square of
the basic variable $u$, which is identical to the convection term present in the Euler equations. First, we note that $u$ is constant along the characteritic curves. Indeed, by the chain rule
\begin{equation}
    f(u)_x = f'(u) u_x = uu_x.
\end{equation}
Using the method of characteristics, for a curve $x(t)$ in the $(x, t)$ plane, the solution is given by $u(x(t), t)$. The total derivative is
\begin{equation}
    \frac{du}{dt} = \frac{\partial u}{\partial t} + \frac{dx}{dt}\frac{\partial u}{\partial x}.
\end{equation}
Comparing this directly to the PDE, we can see that if we choose the characteristic curves that satisfy
\begin{equation}
    \frac{dx}{dt} = 0,
\end{equation}
then along this these curves the total derivative of $u$ satisfies
\begin{equation}
    \frac{du}{dt} = 0.
\end{equation}
This means that $u$ is constant along the characteristic curves which are straight lines.

 A shock forms when these straight-line characteristic curves intersect. The equation of a characteristic curve starting at $(x_0, 0)$ is given by integrating $\frac{dx}{dt} = u$. Since u is constant along this line and equal to its initial value, $u_0(x_0) = u(x_0, 0)$, the equation of the line is
 \begin{equation}
    x(t) = x_0 + u_0(x_0)t.
 \end{equation}
 Consider two distinct characteristic curves starting from points $x_1 < x_2$ at $t=0$. These will intersect if and only if the characteristic starting from behind is 'faster' than the characteritic starting in front, i.e.,
 \begin{equation}
    u_0(x_1) > u_0(x_2).
 \end{equation}
 Thus, for a shock to form $u_0(x)$ must be a decreasing function of $x$ over some interval or equivalently, $u_0'(x) < 0$ for some value of $x$.

 1. Compressive flow: $u_0(x) = \mathbb{1}_{\{x < 0\}}$. For $x < 0$, the characteristic curves are lines with velocity $1$, while for $x > 0$, they are vertical lines ($x = x_0$) with velocity $u=0$. The faster moving curves from the left intersect the stationary characteristics on the right. They all intersect along the line $x = t$ for $t > 0$ forming a shock wave.

 2. Expansive flow: $u_0(x) = \mathbb{1}_{\{x > 0\}}$. For $x < 0$, the characteristics are vertical lines with velocity $u=0$, whereas for $x > 0$, the characteristics have velocity $u=1$. The characteristic now are moving away from each other and do not intersect.

 3. Rarefaction wave: $u_0(x) = \begin{cases}u_1 & x < 0, \\ u_2 & x > 0.\end{cases}$ with $u_1 < u_2$.

    For $x < 0$, the characteristic lines originating from this region are given by $x = x_0 + u_1t$. These are a family of parallel straight lines with velocity $u_1$. The last of these lines emanates from the origin and follows the path $x = u_1t$.

    For $x > 0$, the characteristic lines originating from this region are given by $x = x_0 + u_2t$. These are a family of parallel straight lines with velocity $u_2$. The first of these lines emanates from the origin and follows the path $x = u_2t$.

    Since $u_1 < u_2$, the characteristics originating from the left side of the origin move slower than the characteristics originating from the right side. This means they are diverging, creating a wedge-shaped gap where no characteristic lines from the initial condition exist.
    
    This gap must be filled with a continuous solution known as a rarefaction wave or an expansion fan. This is a self-similar solution that depends only on the ratio $x/t$. By substituting $u(x,t) = g(x/t)$ into the Burger's equation, we find the solution within this fan is $u(x, t) = x/t$.
    
    This solution smoothly connects the state $u_1$ to the state $u_2$. The fan exists in the region between the slowest and fastest moving characteristics that originate from the discontinuity at $x=0$. This region is defined by $u_1t ≤ x ≤ u_2t$.

    Therefore, the complete solution for t > 0 is a three-part piecewise function
    \begin{equation}
        u(x, t) =
        \begin{cases}
            u_1 &x < u₁t, \\
            x/t &u_1t \leq x \leq u_2t, \\
            u_2 &x > u_2t.
        \end{cases}
    \end{equation}
    No shock will form in this case as the characteristic curves on the left move slower than those on the right and the lines diverge from the initial discontinuity at $x = 0$.
