The Airy functions $Ai(z)$ and $Bi(z)$, where $z$ is a complex variable, are two linearly independent solutions of the Airy differential equation
\begin{equation}
    \frac{d^2}{dz^2} y(z) = zy(z),
\end{equation}
satisfying
\begin{equation}
    Ai(0) = \alpha, \quad Ai'(0) = -\beta, \quad Bi(0) = \sqrt{3} \alpha, \quad Bi'(0) = \sqrt{3} \beta,
\end{equation}
where
\begin{equation}
    \alpha = \frac{1}{3^{2/3} \Gamma(\frac{2}{3})} \approx 0.355028053887817, \quad \beta = \frac{1}{3^{1/3} \Gamma(\frac{1}{3})}\approx 0.258819403792807.
\end{equation}
Here $\Gamma$ is the Gamma function, defined by
\begin{equation}
    \Gamma(z) = \int_0^\infty e^{-t}t^{z-1}\,dt.
\end{equation}
The Airy functions are useful in many problems involving transition regions of all kinds, for example in optical diffraction, wave theory, electron
tunnelling, and asymptotic analysis. Also, $Ai$ and $Bi$ have Maclaurin series given by
\begin{equation}
    Ai(z) = \alpha f(z) - \beta g(z), \quad Bi(z) = \sqrt{3}(\alpha f(z) + \beta g(z))
\end{equation}
where
\begin{equation}
    f(z) = 1 + \frac{1}{3!}z^3 + \frac{1 \cdot 4}{6!}z^6 + \frac{1 \cdot 4 \cdot 7}{9!}z^9 + \cdots
\end{equation}
and
\begin{equation}
    g(z) = z + \frac{2}{4!}z^4 + \frac{2 \cdot 5}{7!}z^7 + \frac{2 \cdot 5 \cdot 8}{10!}z^{10} + \cdots
\end{equation}
are power expansions. For large $|z|$, any solution $y(z)$ is given asymptotically by the relation
\begin{equation}
    y(z) \sim AF(z) + BG(z)
\end{equation}
where $A$ and $B$ are complex constants, and where
\begin{equation}
    F(z) = \frac{1}{\sqrt{\pi}}z^{-1/4} \exp\left(-\frac{2}{3}z^{3/2}\right)\left(1 - \frac{5}{48}z^{-3/2} + \cdots\right)
\end{equation}
and
\begin{equation}
    G(z) = \frac{1}{\sqrt{\pi}}z^{-1/4} \exp\left(\frac{2}{3}z^{3/2}\right)\left(1 + \frac{5}{48}z^{-3/2} + \cdots\right)
\end{equation}
where the principal value is taken for any multi-valued function. The values of the constants $A$ and $B$ depend on precisely which solution $y$ is being considered as $Ai$ and $Bi$ have different asymptotic behaviour. Also, they may also depend on which region of the complex plane is under consideration. This is known as Stokes' phenomenon, and the rays from the origin that divide the complex plane into different regions are known as Stokes lines. In the current case, there are three Stokes lines, two of which are given by the rays $\arg{z} = \pm\pi/3$.

In this project, we shall concentrate to start with on the region $R$ given by $|\arg{z}| < \pi/3$. In that region, the appropriate values of $A$ and $B$ are $1/2$ and $0$ respectively for $Ai(z)$. For $Bi(z)$, we need $B = 1$ but $A$ is not important and can take any value (because $F(z)$ is negligible compared to $G(z)$ for large $|z|$ in $R$. Hence $|Ai(z)| \to 0$ and $|Bi(z)| \to \infty$ as $|z| \to \infty$ in $R$.


We claim that
\begin{equation}
    y(z) = \frac{1}{2\pi i} \int_C \exp\left(zt - \frac{1}{3}t^3\right) \,dt
\end{equation}
is a solution of the Airy differential equation, where $C$ is any contour that starts at $\infty e^{-2\pi i/3}$ and ends at $\infty e^{2\pi i/3}.

Differentiating under the integral sign,
\begin{equation}
    y'(z) = \frac{d}{dz} \frac{1}{2\pi i} \int_C \exp(zt - t^3/3)\,dt = \frac{1}{2\pi i} \int_C t\exp(zt - t^3/3)\,dt,
\end{equation}
and differentiating a second time,
\begin{equation}
    y''(z) = \frac{d}{dz} \frac{1}{2\pi i} \int_C t\exp(zt - t^3/3)\,dt = \frac{1}{2\pi i} \int_C t^2\exp(zt - t^3/3)\,dt.
\end{equation}
Now compute
\begin{equation}
    y''(z) - zy(z) = \frac{1}{2\pi i} \int_C (t^2 - z)\exp(zt - t^3/3)\,dt.
\end{equation}
We can observe that the integrand is related to a derivative with respect to $t$,
\begin{equation}
    \frac{d}{dt}\exp(zt - t^3/3) = (z - t^2)\exp(zt - t^3/3) = -(t^2 - z)\exp(zt - t^3/3).
\end{equation}
Therefore, the integral becomes
\begin{equation}
    y''(z) - zy(z) = -\frac{1}{2\pi i} \int_C \frac{d}{dt} \exp(zt - t^3/3)\,dt
\end{equation}
evaluated at the endpoints of $C$. The dominant term in the exponent for large $|t|$ is $-t^3/3$. Let $t = Re^{2\pi i/3}$. Then $t^3 = R^3$ so the dominant tends to $-\infty$ as $R \to \infty$, hence the exponent itself tends to zero. Now let $t = Re^{-2\pi i/3}$. Then also $t^3 = R^3$. The dominant term again tends to $-\infty$ as $R \to \infty$. Thus, the exponent also tends to zero. Since the integral evaluates to zero at both endpoints, by the fundamental theorem of contour integrals, the entire integral is zero. This confirms that $y(z)$ is a solution to the Airy equation.

Setting $z = 0$ in the expression for $y(z)$ yields
\begin{equation}
    y(0) = \frac{1}{2\pi i} \int_C \exp(-t^3/3)\,dt.
\end{equation}
We deform the contour $C$ into two straight rays, one from $\infty e^{-2\pi i/3}$ to the origin denoted $C_1$, and the other from the origin to $\infty e^{2\pi i/3}$ denoted $C_2$. For $C_2$, we parameterise with $t = re^{2\pi i/3}$, where $r$ goes from $0$ to $\infty$. Then $dt = e^{2\pi i/3}\,dr$ and $t^3 = r^3$. Similarly, for $C_1$, we parameterise with $t = re^{-2\pi i/3}$, where $r$ goes from $\infty$ to $0$. Then $dt = e^{-2\pi i/3}\,dr$ and $t^3 = r^3$. Hence
\begin{align}
    y(0)
    &= \frac{1}{2\pi i} (e^2\pi i/3 - e^{-2\pi i/3}) ∫_0^\infty \exp(-r^3/3) \,dr \\
    &= \frac{1}{2\pi i}2i\sin(2\pi/3) ∫_0^\infty \exp(-r^3/3)\,dr = \frac{1}{\pi} \frac{\sqrt{3}}{2} \int_0^\infty \exp(-r^3/3)\,dr
\end{align}

Let $u = r^3/3$, so $r = (3u)^{1/3}$ and $dr = (3u)^{-2/3}\,du$. Then
\begin{equation}
    \int_0^\infty \exp(-r^3/3)\,dr = \int_0^\infty e^{-u}(3u)^{-2/3}\,du = 3^{-2/3} \int_0^\infty u^{-2/3}e^{-u}\,du = 3^{-2/3}\Gamma\left(\frac{1}{3}\right).
\end{equation}
Thus, using the reflection formula $Γ(z)Γ(1-z) = \pi/\sin(\pi z)$,
\begin{equation}
    y(0) = \frac{\sqrt{3}}{2\pi} 3^{-2/3} \Gamma\left(\frac{1}{3}\right) = \frac{1}{3^{2/3}\Gamma(\frac{2}{3})} = \alpha.
\end{equation}

Now, setting $z = 0$ in the expression for $y'(z)$, using the same deformed contour, same substitution and reflection formula, we obtain
\begin{align}
    y'(0)
    &= \frac{1}{2\pi i} \int_C t\exp(-t^3/3)\,dt \\
    &= \frac{1}{\pi}\frac{-\sqrt{3}}{2} ∫_0^\infty r\exp(-r^3/3)\,dr \\
    &= \frac{1}{\pi}\frac{-\sqrt{3}}{2} 3^{-1/3} \int_0^\infty u^{-1/3}e^{-u}\,du \\
    &= \frac{1}{\pi}\frac{-\sqrt{3}}{2} 3^{-1/3} \Gamma\left(\frac{2}{3}\right) \\
    &= -\frac{1}{3^{1/3}\Gamma(\frac{1}{3})} = -\beta.
\end{align}
By the existence and uniqueness theorem of initial value problems, it must be the case that $y(z) = Ai(z)$.
