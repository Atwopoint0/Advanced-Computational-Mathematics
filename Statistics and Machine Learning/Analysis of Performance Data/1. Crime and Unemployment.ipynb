Let $n_{1i}$ and $n_{2i}$ be the numbers of offences committed by the $i$-th boy for $i \in \{1, \dots, 32\}$ in Employment and Unemployment respectively, in times $t_{1i}$ and $t_{2i}$ years. Assume that $n_{1i}$ and $n_{2i}$ are independent Poisson variables, with parameters $\lambda_{i}t_{1i}^\alpha$ and $\phi\lambda_it_{2i}^\alpha$
respectively. Here $\alpha$ and $\phi$ represent the parameters of interest with $\phi$ being the additional risk (if any) of committing an offence while
unemployed rather than while employed.

The probability mass function for a Poisson variable $X$ with parameter $\mu$ is
\begin{equation}
    \Pr(X = k) = \frac{e^{-\mu}\mu^k}{k!}.
\end{equation}
Since $n_{1i}$ and $n_{2i}$ are independent, the joint probability for a specific boy $i$ is the product of their individual probabilities. Substituting the Poisson PMFs yields
\begin{equation}
    \Pr(\text{data}_i) = \frac{e^{-\lambda_i t_{1i}^\alpha} (\lambda_i t_{1i}^\alpha)^{n_{1i}}}{n_{1i}!} Â· \frac{e^{-\phi \lambda_i t_{2i}^\alpha} (\phi \lambda_i t_{2i}^\alpha)^{n_{2i}}}{n_{2i}!}.
\end{equation}
The total likelihood $L$ for the entire data set (all $32$ boys) is the product of the individual probabilities, assuming the observations for different boys are independent,
\begin{align}
    L(\phi, \alpha, \mathbf{\lambda}) &= \prod_{i=1}^{32} \left( \frac{e^{-\lambda_i t_{1i}^\alpha} (\lambda_i t_{1i}^\alpha)^{n_{1i}}}{n_{1i}!} \cdot \frac{e^{-\phi \lambda_i t_{2i}^\alpha} (\phi \lambda_i t_{2i}^\alpha)^{n_{2i}}}{n_{2i}!} \right) \\
    &= \prod_{i=1}^{32} \left[ \frac{1}{n_{1i}! n_{2i}!} \phi^{n_{2i}} \lambda_i^{(n_{1i} + n_{2i})} t_{1i}^{\alpha n_{1i}} t_{2i}^{\alpha n_{2i}} \exp({-\lambda_i (t_{1i}^\alpha + \phi t_{2i}^\alpha)}) \right].
\end{align}

---

We want to now find the conditional distribution of $n_{2i}$ given the sum $m_i = n_{1i} + n_{2i}$. Recall from probability theory that if $X \sim \text{Poisson}(\mu_1)$ and $Y \sim \text{Poisson}(\mu_2)$ are independent, then their conditional distribution given their sum $X+Y=m$ is Binomial with parameters $m$ and success probability $p = \frac{\mu_1}{\mu_1 + \mu_2}$ for $X$ or $p = \frac{\mu_2}{\mu_1 + \mu_2}$ for $Y$. Hence, in our case
\begin{equation}
    \Pr(n_{2i} = k \mid n_{1i} + n_{2i} = m_i) = \frac{\Pr(n_{2i} = k \cap n_{1i} = m_i - k)}{\Pr(n_{1i} + n_{2i} = m_i)}
\end{equation}
Since $n_{1i}$ and $n_{2i}$ are independent, the numerator is the product of their individual probabilities. The denominator is the probability of a sum of Poisson variables, which is itself a Poisson variable with mean $\mu_{1i} + \mu_{2i}$,
\begin{align}
    \Pr(n_{2i} = k \mid n_{1i} + n_{2i} = m_i) &= \frac{\left( \frac{e^{-\mu_{2i}}\mu_{2i}^k}{k!} \right) \left( \frac{e^{-\mu_{1i}}\mu_{1i}^{m_i-k}}{(m_i-k)!} \right)}{ \frac{e^{-(\mu_{1i}+\mu_{2i})}(\mu_{1i}+\mu_{2i})^{m_i}}{m_i!} } \\
    &= \binom{m_i}{k} \left( \frac{\mu_{2i}}{\mu_{1i} + \mu_{2i}} \right)^k \left( \frac{\mu_{1i}}{\mu_{1i} + \mu_{2i}} \right)^{m_i-k}.
\end{align}
This is precisely the probability mass function of a Binomial distribution with parameters $m_i$ and probability $\psi_i$, where
\begin{equation}
    \psi_i = \frac{\mu_{2i}}{\mu_{1i} + \mu_{2i}}.
\end{equation}

Now we substitute the definitions of $\mu_{1i}$ and $\mu_{2i}$ into the expression for $\psi_i$ and $1 - \psi_i$,
\begin{align}
    \psi_i &= \frac{\phi \lambda_i t_{2i}^\alpha}{\lambda_i t_{1i}^\alpha + \phi \lambda_i t_{2i}^\alpha} = \frac{\phi t_{2i}^\alpha}{t_{1i}^\alpha + \phi t_{2i}^\alpha}, \\
    1 - \psi_i &= 1 - \frac{\phi t_{2i}^\alpha}{t_{1i}^\alpha + \phi t_{2i}^\alpha} = \frac{t_{1i}^\alpha}{t_{1i}^\alpha + \phi t_{2i}^\alpha}.
\end{align}
Now, taking the ratio gives
\begin{equation}
    \frac{\psi_i}{1 - \psi_i} = \frac{\phi t_{2i}^\alpha}{t_{1i}^\alpha} = \phi \left( \frac{t_{2i}}{t_{1i}} \right)^\alpha,
\end{equation}
and finally taking logarithm of both sides yields the log-odds
\begin{equation}
    \log \left( \frac{\psi_i}{1 - \psi_i} \right) = \log \phi + \alpha \log \left( \frac{t_{2i}}{t_{1i}} \right).
\end{equation}

The likelihood of the data conditional on the totals $m_i$ is the product of the individual binomial probabilities for all $32$ boys. We can directly compute this:
\begin{align}
    L(\phi, \alpha)
    &= \prod_{i=1}^{32} P(n_{2i} \mid m_i) \\
    &= \prod_{i=1}^{32} \binom{m_i}{n_{2i}} \psi_i^{n_{2i}} (1-\psi_i)^{m_i - n_{2i}} \\
    &= \prod_{i=1}^{32} \binom{m_i}{n_{2i}} \left( \frac{\phi t_{2i}^\alpha}{t_{1i}^\alpha + \phi t_{2i}^\alpha} \right)^{n_{2i}} \left( \frac{t_{1i}^\alpha}{t_{1i}^\alpha + \phi t_{2i}^\alpha} \right)^{n_{1i}} \\
    &= \prod_{i=1}^{32} \binom{m_i}{n_{2i}} \frac{\phi^{n_{2i}} t_{2i}^{\alpha n_{2i}} t_{1i}^{\alpha n_{1i}}}{(t_{1i}^\alpha + \phi t_{2i}^\alpha)^{m_i}}.
\end{align}
This is identical to a logistic regression model where the dependent variable is the number of successes $n_{2i}$ out of $m_i$ trials, the intercept parameter is $\beta_0 = \log \phi$, the slope parameter is $\beta_1 = \alpha$ and the covariate (predictor) is $x_i = \log(t_{2i} / t_{1i})$. We will use the Newton-Raphson algorithm to find estimates for $\beta_0$ and $\beta_1$.

```python
# @title Data
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import io

data_str = """
Boy,    Em (Yr),    Offences Em,    Un (Yr),    Offences Un
1,      0.83,       1,              0.88,       3
2,      1.02,       0,              0.89,       8
3,      1.17,       0,              2.5,        2
4,      1.21,       3,              0.68,       1
5,      1.25,       4,              0.96,       2
6,      1.31,       2,              0.69,       1
7,      1.5,        2,              1.25,       0
8,      1.52,       2,              0.64,       2
9,      1.54,       1,              0.85,       0
10,     1.58,       2,              1.08,       0
11,     1.66,       4,              0.61,       1
12,     1.75,       1,              0.25,       0
13,     2.21,       1,              1.37,       4
14,     2.22,       0,              0.28,       1
15,     2.25,       0,              0.75,       1
16,     2.37,       1,              0.55,       0
17,     2.39,       2,              0.44,       1
18,     2.42,       3,              0.78,       0
19,     2.45,       1,              0.47,       0
20,     2.51,       2,              0.4,        0
21,     2.64,       1,              0.7,        0
22,     2.65,       8,              0.6,        0
23,     2.67,       1,              0.67,       0
24,     2.83,       1,              0.37,       0
25,     2.84,       1,              0.75,       0
26,     2.97,       2,              0.36,       0
27,     3,          1,              0.34,       0
28,     3.07,       1,              0.27,       0
29,     3.15,       2,              0.43,       0
30,     3.16,       1,              0.67,       1
31,     3.28,       4,              0.45,       0
32,     3.37,       2,              0.3,        0
"""
```

```python
def load_and_preprocess_data(csv_string):
    '''
    Parses the CSV string and constructs the design matrix X
    and response vectors needed for the Binomial model.
    '''
    df = pd.read_csv(io.StringIO(csv_string))
    # Strip whitespace from column names
    df.columns = df.columns.str.strip()
    n1 = df['Offences Em'].values
    n2 = df['Offences Un'].values
    t1 = df['Em (Yr)'].values
    t2 = df['Un (Yr)'].values

    # Derived variables for the binomial model
    # y: successes (offences while unemployed)
    # m: trials (total offences)
    y = n2
    m = n1 + n2

    # Predictor variable: log(t2 / t1)
    # The model is log(psi / 1-psi) = log_phi + alpha * log(t2/t1)
    x_feature = np.log(t2 / t1)

    # Design natrix X: Column of 1s (intercept) and column of x_feature
    X = np.column_stack((np.ones(len(x_feature)), x_feature))

    return X, y, m
```

```python
def sigmoid(z):
    '''
    Computes the sigmoid function safely.
    '''
    return 1 / (1 + np.exp(-z))

def compute_log_likelihood(beta, X, y, m):
    '''
    Calculates the log-likelihood for the binomial model.
    beta: array of parameters [intercept, slope]
    '''
    theta = X @ beta
    psi = sigmoid(theta)

    # Add epsilon to avoid log(0) error
    epsilon = 1e-9

    # Log likelihood = Sum [ y*ln(psi) + (m-y)*ln(1-psi) ]
    ll_terms = y * np.log(psi + epsilon) + (m - y) * np.log(1 - psi + epsilon)
    return np.sum(ll_terms)

def compute_gradient_and_hessian(X, y, m, psi):
    '''
    Computes the gradient (Score) and hessian (Information) matrices.
    '''
    # Gradient: X.T * (y - m * psi)
    gradient = X.T @ (y - m * psi)

    # Hessian: -X.T * W * X, where W is diagonal matrix of weights
    weights = m * psi * (1 - psi)

    # Efficient calculation of X.T @ W @ X
    # We multiply each row of X by the corresponding weight
    WX = X * weights[:, np.newaxis]
    hessian = -X.T @ WX

    return gradient, hessian

def run_newton_raphson(X, y, m, max_iter=20, tol=1e-6):
    '''
    Performs the Newton-Raphson optimization to find MLE.
    Returns beta_hat and the variance-covariance matrix
    '''
    # Initialise parameters (beta = [0, 0])
    beta = np.zeros(X.shape[1])

    for i in range(max_iter):
        theta = X @ beta
        psi = sigmoid(theta)

        gradient, hessian = compute_gradient_and_hessian(X, y, m, psi)

        try:
            inv_hessian = np.linalg.inv(hessian)
        except np.linalg.LinAlgError:
            raise ValueError("Hessian matrix is singular.")

        # Newton update step: beta_new = beta_old - H^-1 * gradient
        delta = inv_hessian @ gradient
        beta = beta - delta

        # Check convergence
        if np.sum(np.abs(delta)) < tol:
            break

    # The Covariance Matrix of the MLE is -H^-1 (inverse of observed information)
    # Note: inv_hessian is H^-1. So we return -inv_hessian.
    cov_matrix = -inv_hessian

    return beta, cov_matrix
```

```python
def calculate_estimates_and_errors(beta_hat, cov_matrix):
    '''
    Extracts parameter estimates and calculates standard errors,
    including the delta method transformation for phi.
    '''
    # Extract raw estimates
    log_phi_hat = beta_hat[0]
    alpha_hat = beta_hat[1]

    # Transform log_phi back to phi
    phi_hat = np.exp(log_phi_hat)

    # Extract standard errors for beta parameters (sqrt of diagonal of covariance)
    se_log_phi = np.sqrt(cov_matrix[0, 0])
    se_alpha = np.sqrt(cov_matrix[1, 1])

    # Delta method for SE of phi
    # Var(phi) approx [d(exp(u))/du]^2 * Var(u) = phi^2 * Var(log_phi)
    se_phi = phi_hat * se_log_phi

    results = {
        "alpha": alpha_hat,
        "se_alpha": se_alpha,
        "log_phi": log_phi_hat,
        "se_log_phi": se_log_phi,
        "phi": phi_hat,
        "se_phi": se_phi
    }
    return results

def print_results(res):
    '''
    Formatted printing of the results.
    '''
    print("-" * 40)
    print("Maximum Likelihood Estimates")
    print("-" * 40)
    print(f"Alpha (slope)      : {res['alpha']:.4f} (SE: {res['se_alpha']:.4f})")
    print(f"Log Phi (intercept): {res['log_phi']:.4f} (SE: {res['se_log_phi']:.4f})")
    print(f"Phi (transformed)  : {res['phi']:.4f} (SE: {res['se_phi']:.4f})")
    print("-" * 40)

def plot_conditional_likelihood(X, y, m, beta_hat):
    '''
    Plots the conditional log-likelihood as a function of log_phi,
    holding alpha constant at its MLE value.
    '''
    log_phi_hat = beta_hat[0]
    alpha_hat = beta_hat[1]

    # Define a range around the MLE for log_phi
    x_range = np.linspace(log_phi_hat - 1.5, log_phi_hat + 1.5, 100)
    ll_values = []

    for val in x_range:
        # Construct a temporary beta vector [varying_log_phi, fixed_alpha]
        temp_beta = np.array([val, alpha_hat])
        ll = compute_log_likelihood(temp_beta, X, y, m)
        ll_values.append(ll)

    plt.figure(figsize=(8, 5))
    plt.plot(x_range, ll_values, label=r'Log-Likelihood ($\alpha$ fixed)')
    plt.axvline(log_phi_hat, color='red', linestyle='--',
                label=f'MLE $\\log \\phi$ = {log_phi_hat:.3f}')

    plt.title(r'Profile Log-Likelihood for $\log \phi$')
    plt.xlabel(r'$\log \phi$')
    plt.ylabel('Log-Likelihood')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
X, y, m = load_and_preprocess_data(data_str)
beta_hat, cov_matrix = run_newton_raphson(X, y, m)
stats = calculate_estimates_and_errors(beta_hat, cov_matrix)
print_results(stats)
plot_conditional_likelihood(X, y, m, beta_hat)
```

The value of $\phi$ represents the additional risk of offending while unemployed compared to employed. From our data, conditional on the total number of offenses, the rate of offending is roughly $\hat\phi$ times higher during periods of unemployment than employment, adjusting for the duration. Looking at the standard errors, $\phi = 1$ is outside two standard deviations of $\hat\phi$.

To determine if this result is statistically significant we must test the null hypothesis $H_0: \log \phi = 0$ against the alternative $H_1: \log \phi \neq 0$. We use the Wald test ($Z$-score)
\begin{equation}
    Z = \frac{\log \hat\phi - 0}{\text{SE}(\log \hat\phi)},
\end{equation}
which in our case is greater than the critical value $1.96$ for the $95\%$ confidence interval for a two-tailed test. We reject the null hypothesis at the $5\%$ significance level. The approximate $95\%$ confidence interval for $\log \phi$ excludes $0$.

The log-likelihood as a function of $\log \phi$ with $\alpha$ fixed is an inverted parabola whose peack occurs at the MLE value $\log \hat\phi$. The shape is smooth, unimodal, and approximately quadratic near the peak. This quadratic behavior is expected for log-likelihood functions of exponential family distributions near their maximum. The width of the parabola relates to the information about the parameter; a sharper peak would indicate lower variance, while a flatter peak indicates higher variance.
